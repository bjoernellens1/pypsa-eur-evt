```python
# def read_and_standardize(file_path, source_name, tz="Europe/Vienna"):
#     """
#     Reads a semicolon-delimited CSV, parses datetime index, and standardizes the format.
#     Converts all values from GW to MW (PyPSA uses MW internally).
#     Unit system: input = GW, output = MW
#     """
#     logging.info(f"Reading {source_name} from {file_path}")

#     df = pd.read_csv(file_path, sep=";", engine="python")

#     # Detect datetime column
#     datetime_column = None
#     for col in df.columns:
#         if "date" in col.lower() or "time" in col.lower() or "zeitpunkt" in col.lower():
#             datetime_column = col
#             break

#     if datetime_column is None:
#         raise ValueError(f"No datetime column found in {file_path}")

#     df[datetime_column] = pd.to_datetime(df[datetime_column], dayfirst=True)
#     df.set_index(datetime_column, inplace=True)

#     # Ensure timezone-awareness
#     df.index = df.index.tz_localize(tz, ambiguous='NaT', nonexistent='shift_forward')

#     # Convert all columns to numeric and from GW to MW
#     df = df.apply(pd.to_numeric, errors='coerce') * 1000

#     # Clean column names
#     df.columns = [str(c).strip() for c in df.columns]

#     df.sort_index(inplace=True)

#     return df

def read_and_standardize(file_path, source_name, tz="Europe/Vienna"):
    """
    Reads a semicolon-delimited CSV, parses datetime index, and standardizes the format.
    Converts all values from GW to MW (PyPSA uses MW internally).
    Unit system: input = GW, output = MW
    """
    logging.info(f"Reading {source_name} from {file_path}")

    df = pd.read_csv(file_path, sep=";", engine="python")

    # Detect datetime column
    datetime_column = None
    for col in df.columns:
        if "date" in col.lower() or "zeit" in col.lower() or "time" in col.lower():
            datetime_column = col
            break
    if datetime_column is None:
        datetime_column = df.columns[0]

    df[datetime_column] = pd.to_datetime(df[datetime_column], utc=True)
    df = df.set_index(datetime_column)
    df = df.tz_convert("UTC")
    # Convert all columns to MW (from GW)
    for col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce") * 1000.0
    df = df.squeeze()
    return df
```

```python
def resample_hourly(df, hourly_index, method="interpolate"):
    """
    Resample or reindex a DataFrame or Series to the provided hourly_index.
    Fills missing values using interpolation or forward/backward fill as needed.
    """
    df = ensure_unique_index(df)
    df = df.reindex(hourly_index)
    if method == "interpolate":
        df = df.interpolate(method="time")
    elif method == "ffill":
        df = df.ffill()
    elif method == "bfill":
        df = df.bfill()
    return df
```

```python
def ensure_unique_index(df):
    """Ensure the index is unique before reindexing."""
    return df[~df.index.duplicated(keep='first')]
```
