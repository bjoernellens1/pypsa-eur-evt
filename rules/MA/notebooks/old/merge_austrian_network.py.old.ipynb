```python
import pandas as pd
import numpy as np
from scipy.spatial import cKDTree

# Copy bus DataFrames
buses1 = network1.buses.copy()
buses2 = networkAT.buses.copy()

# Ensure type column is numeric
buses2["type"] = pd.to_numeric(buses2["type"], errors="coerce")

# Select Austrian buses
austrian_buses1 = buses1[buses1["country"].isin(["AT", "Austria"])].copy()

# Select reference buses (type 3 and v_nom 220 or 380 kV)
reference_buses2 = buses2[
    (buses2["type"] == 3) & 
    (buses2["v_nom"].isin([220, 380]))
].dropna(subset=["x", "y"]).copy()

if reference_buses2.empty:
    raise ValueError("No valid reference buses found.")

# Build KDTree on Austrian buses
tree_at = cKDTree(austrian_buses1[["x", "y"]].values)

# Query each reference bus's nearest Austrian bus
dists, at_indices = tree_at.query(reference_buses2[["x", "y"]].values)

# Deduplicate: one-to-one mapping from Austrian buses to reference buses
assigned_at_buses = set()
updates = []

for ref_idx, at_idx in enumerate(at_indices):
    if at_idx not in assigned_at_buses:
        assigned_at_buses.add(at_idx)
        ref_coords = reference_buses2.iloc[ref_idx][["x", "y"]]
        at_bus_idx = austrian_buses1.index[at_idx]
        updates.append((at_bus_idx, ref_coords["x"], ref_coords["y"]))

# Apply coordinate updates
for bus_idx, new_x, new_y in updates:
    buses1.at[bus_idx, "x"] = new_x
    buses1.at[bus_idx, "y"] = new_y

# Build merged network
merged_network = network1.copy()
merged_network.buses.loc[buses1.index, ["x", "y"]] = buses1[["x", "y"]]

# Summary
print(f"Updated {len(updates)} of {len(austrian_buses1)} Austrian buses.")
print(f"Left {len(austrian_buses1) - len(updates)} Austrian buses in original position.")
```

```python
import pandas as pd
import numpy as np
from scipy.spatial import cKDTree
import pypsa # Ensure PyPSA is installed

def calculate_distance(x1, y1, x2, y2, coords_are_lat_lon=False):
    """
    Calculate distance between two points.
    If coords_are_lat_lon is True, uses Haversine formula (approximate).
    Otherwise, uses Euclidean distance (assuming projected coordinates).
    Returns distance in km if lat/lon, or in native units if projected.
    """
    if coords_are_lat_lon:
        R = 6371  # Earth radius in kilometers
        lat1_rad = np.radians(y1)
        lon1_rad = np.radians(x1)
        lat2_rad = np.radians(y2)
        lon2_rad = np.radians(x2)
        dlon = lon2_rad - lon1_rad
        dlat = lat2_rad - lat1_rad
        a = np.sin(dlat / 2)**2 + np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(dlon / 2)**2
        c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
        distance = R * c
        return distance
    else:
        # Assuming x, y are in a projected coordinate system (e.g., meters)
        # To return in km, divide by 1000 if original units are meters
        return np.sqrt((x1 - x2)**2 + (y1 - y2)**2) # / 1000 if x,y in meters for km


def merge_european_and_austrian_networks(network1: pypsa.Network, networkAT: pypsa.Network,
                                         coords_are_lat_lon: bool = False,
                                         convert_distance_to_km: bool = False,
                                         projected_crs_unit_to_km: float = 1.0):
    """
    Merges network1 (e.g., PyPSA-Eur) with networkAT (detailed Austrian network).

    Args:
        network1: PyPSA Network object (larger, less detailed).
        networkAT: PyPSA Network object (smaller, more detailed, e.g., Austria).
        coords_are_lat_lon: Set to True if bus 'x' and 'y' are lon and lat.
        convert_distance_to_km: If True and coords are not lat/lon, applies projected_crs_unit_to_km.
        projected_crs_unit_to_km: Factor to convert Euclidean distance to km (e.g., 1/1000 if x,y in meters).
    Returns:
        A new PyPSA Network object representing the merged network.
    """

    print("Starting network merging process...")

    # 1. DATA PREPARATION AND INITIAL SELECTION
    buses1_df = network1.buses.copy()
    buses2_df = networkAT.buses.copy()

    if not isinstance(buses2_df.index, pd.RangeIndex): # Ensure index is usable for iloc later if needed
        buses2_df = buses2_df.reset_index() # Keep original name if it's the index
        # If 'name' or 'index' was the original index, ensure it's a column now. Often PyPSA uses 'name' as index.
        # This script assumes bus names are in `buses_df.index`. If not, adjust accordingly.

    # Ensure 'type' column is numeric for filtering, coercing errors
    if "type" in buses2_df.columns:
        buses2_df["type"] = pd.to_numeric(buses2_df["type"], errors="coerce")
    else: # If type column is missing, we cannot filter by it.
        print("Warning: 'type' column missing in networkAT.buses. Cannot filter reference buses by type.")


    # Select Austrian buses from network1
    austrian_buses1_df = buses1_df[buses1_df.get("country", pd.Series(dtype=str)).isin(["AT", "Austria"])].copy()
    if austrian_buses1_df.empty:
        print("Warning: No Austrian buses found in network1 based on 'country' codes 'AT' or 'Austria'.")
        # Depending on desired behavior, could raise error or proceed with only networkAT additions

    # Select reference buses from networkAT (e.g., type 3, high voltage substations)
    # Adjust filter as needed for 'type 3' definition and voltage levels
    type_filter = (buses2_df.get("type", pd.Series(dtype=float)) == 3) if "type" in buses2_df.columns else pd.Series(True, index=buses2_df.index)
    v_nom_filter = buses2_df.get("v_nom", pd.Series(dtype=float)).isin([220, 380, 220.0, 380.0]) # Allow float v_nom

    reference_buses2_df = buses2_df[
        type_filter & v_nom_filter
    ].dropna(subset=["x", "y"]).copy()

    if reference_buses2_df.empty:
        print("Warning: No valid reference buses (type 3, 220/380kV with coordinates) found in networkAT.")
        # If no reference buses, we might only add them if they exist, or skip parts of the logic.
        # For now, we'll let it proceed; unmerged logic might still pick them up if any.
    else:
        print(f"Selected {len(reference_buses2_df)} reference buses from networkAT.")


    # 2. SPATIAL MATCHING AND PREPARING COORDINATE UPDATES
    # This map will store: original_ref2_bus_name -> final_merged_network_bus_name
    map_ref2_original_to_final_merged_name = {}
    # This set stores original names of networkAT reference buses that are used to update network1 buses
    used_networkAT_ref_bus_original_names = set()
    # This list stores tuples: (network1_bus_name_to_update, new_x_coord, new_y_coord)
    updates_for_network1_coords = []

    if not austrian_buses1_df.empty and not reference_buses2_df.empty:
        tree_at_net1 = cKDTree(austrian_buses1_df[["x", "y"]].values)
        # Query for each reference_buses2, its nearest austrian_buses1
        # `reference_buses2_df.index` here gives the original indices/names from networkAT.buses
        # `austrian_buses1_df.index` gives original indices/names from network1.buses
        ref2_coords = reference_buses2_df[["x", "y"]].values
        if ref2_coords.size > 0:
            distances, at_indices_iloc = tree_at_net1.query(ref2_coords)

            # Deduplicate: ensure one-to-one mapping from austrian_buses1 to reference_buses2
            assigned_at_buses_net1_iloc = set() # Using iloc for austrian_buses1_df

            for i, ref2_bus_original_name in enumerate(reference_buses2_df.index):
                matched_at_bus_iloc = at_indices_iloc[i]
                if matched_at_bus_iloc not in assigned_at_buses_net1_iloc:
                    assigned_at_buses_net1_iloc.add(matched_at_bus_iloc)

                    ref2_bus_coords = reference_buses2_df.iloc[i][["x", "y"]]
                    at_bus_original_name_in_net1 = austrian_buses1_df.index[matched_at_bus_iloc]

                    updates_for_network1_coords.append(
                        (at_bus_original_name_in_net1, ref2_bus_coords["x"], ref2_bus_coords["y"])
                    )
                    used_networkAT_ref_bus_original_names.add(ref2_bus_original_name)
                    map_ref2_original_to_final_merged_name[ref2_bus_original_name] = at_bus_original_name_in_net1
    else:
        print("Skipping spatial matching due to empty Austrian buses in network1 or reference buses in networkAT.")

    print(f"Prepared {len(updates_for_network1_coords)} coordinate updates for existing Austrian buses in network1.")

    # 3. CREATE BASE MERGED_NETWORK AND APPLY INITIAL COORDINATE UPDATES
    # Create a deep copy of network1. If network1 has time-dependent data, decide `with_time`.
    merged_network = network1.copy(with_time=False) # Or True, as needed

    for bus_name, new_x, new_y in updates_for_network1_coords:
        if bus_name in merged_network.buses.index:
            merged_network.buses.at[bus_name, "x"] = new_x
            merged_network.buses.at[bus_name, "y"] = new_y
        else:
            print(f"Warning: Bus {bus_name} for coordinate update not found in merged_network.buses. Skipping.")
    print(f"Applied {len(updates_for_network1_coords)} coordinate updates to merged_network.")


    # 4. IDENTIFY UNMERGED networkAT REFERENCE BUSES
    all_networkAT_ref_bus_original_names = set(reference_buses2_df.index)
    unmerged_networkAT_ref_bus_original_names = all_networkAT_ref_bus_original_names - used_networkAT_ref_bus_original_names
    unmerged_buses_from_networkAT_df = reference_buses2_df.loc[list(unmerged_networkAT_ref_bus_original_names)].copy()
    print(f"Identified {len(unmerged_buses_from_networkAT_df)} reference buses from networkAT to be added as new buses.")

    # 5. ADD UNMERGED BUSES TO MERGED_NETWORK
    if not unmerged_buses_from_networkAT_df.empty:
        for original_name, row_data in unmerged_buses_from_networkAT_df.iterrows():
            new_bus_name_in_merged = f"AT_NET2_{original_name}"
            # Ensure unique name
            suffix_counter = 1
            temp_name = new_bus_name_in_merged
            while temp_name in merged_network.buses.index:
                temp_name = f"{new_bus_name_in_merged}_{suffix_counter}"
                suffix_counter += 1
            new_bus_name_in_merged = temp_name

            merged_network.add(
                "Bus",
                name=new_bus_name_in_merged,
                v_nom=row_data["v_nom"],
                x=row_data["x"],
                y=row_data["y"],
                country="AT",
                type=str(row_data.get("type", "")), # Ensure type is string if it was numeric
                carrier=row_data.get("carrier", "AC"), # Default carrier
                control=row_data.get("control", "AT0") # Default control area
                # Add other relevant bus attributes from row_data if needed
            )
            map_ref2_original_to_final_merged_name[original_name] = new_bus_name_in_merged
        print(f"Added {len(unmerged_buses_from_networkAT_df)} new buses from networkAT to merged_network.")


    # 6. FIND NEAREST NEIGHBORS FOR (NOW ADDED) UNMERGED BUSES (within networkAT reference context)
    new_line_candidates = [] # Stores (original_unmerged_name, original_neighbor_name_from_ref2)

    if not unmerged_buses_from_networkAT_df.empty and len(reference_buses2_df) > 1:
        # KDTree on all reference buses from networkAT to find neighbors among them
        tree_all_ref2_buses = cKDTree(reference_buses2_df[["x", "y"]].values)

        for unmerged_bus_original_name, unmerged_bus_row in unmerged_buses_from_networkAT_df.iterrows():
            query_coords = unmerged_bus_row[["x", "y"]].values.reshape(1, -1)
            # Query for k=2 because the closest (k=1) will be the bus itself.
            # Need to handle if len(reference_buses2_df) is small.
            k_neighbors = min(2, len(reference_buses2_df))
            if k_neighbors < 1: continue # Should not happen if len(reference_buses2_df) > 1

            distances_nn, nn_indices_iloc = tree_all_ref2_buses.query(query_coords, k=k_neighbors)

            if k_neighbors == 1 and reference_buses2_df.index[nn_indices_iloc[0][0]] == unmerged_bus_original_name:
                 # Only itself found, no other neighbor
                print(f"Unmerged bus {unmerged_bus_original_name} could not find a distinct neighbor in reference_buses2_df.")
                continue

            # The first index (0) is the bus itself if k_neighbors > 0
            # The second index (1) is its nearest distinct neighbor (if k_neighbors=2)
            neighbor_iloc_in_ref2 = nn_indices_iloc[0][0] if reference_buses2_df.index[nn_indices_iloc[0][0]] != unmerged_bus_original_name else (nn_indices_iloc[0][1] if k_neighbors > 1 else -1)

            if neighbor_iloc_in_ref2 != -1:
                neighbor_original_name_in_ref2 = reference_buses2_df.index[neighbor_iloc_in_ref2]
                if unmerged_bus_original_name != neighbor_original_name_in_ref2: # Ensure not connecting to self
                    new_line_candidates.append((unmerged_bus_original_name, neighbor_original_name_in_ref2))
    print(f"Identified {len(new_line_candidates)} potential new lines for unmerged networkAT buses.")


    # 7. ADD NEW LINES TO MERGED_NETWORK
    if new_line_candidates:
        num_existing_lines = len(merged_network.lines)
        for i, (orig_unmerged_name, orig_neighbor_name) in enumerate(new_line_candidates):
            bus0_final_name = map_ref2_original_to_final_merged_name.get(orig_unmerged_name)
            bus1_final_name = map_ref2_original_to_final_merged_name.get(orig_neighbor_name)

            if bus0_final_name and bus1_final_name and bus0_final_name != bus1_final_name:
                new_line_name = f"NEW_AT_Line_{num_existing_lines + i}"
                 # Ensure unique name
                suffix_counter = 1
                temp_name = new_line_name
                while temp_name in merged_network.lines.index: # Check against existing line names
                    temp_name = f"{new_line_name}_{suffix_counter}"
                    suffix_counter += 1
                new_line_name = temp_name

                # Get coordinates for length calculation
                bus0_data = merged_network.buses.loc[bus0_final_name]
                bus1_data = merged_network.buses.loc[bus1_final_name]
                
                length_val = calculate_distance(bus0_data['x'], bus0_data['y'], bus1_data['x'], bus1_data['y'],
                                                coords_are_lat_lon)
                if convert_distance_to_km and not coords_are_lat_lon:
                    length_km = length_val * projected_crs_unit_to_km
                elif coords_are_lat_lon: # Already in km
                    length_km = length_val
                else: # In native units of projected CRS
                    length_km = length_val # Use as is, or set a default if interpretation is unclear

                if length_km <= 0: length_km = 0.1 # Avoid zero or negative length for parameters

                # Placeholder typical line parameters per km (adjust based on actual data/voltage)
                v_nom = bus0_data['v_nom'] # Assuming bus0 and bus1 have compatible v_nom for a line
                r_per_km, x_per_km, b_per_km, s_nom_typical = 0.07, 0.35, 2.5e-6, 500 # Default for 220kV
                if v_nom == 380:
                    r_per_km, x_per_km, b_per_km, s_nom_typical = 0.025, 0.25, 4.5e-6, 1500
                elif v_nom < 220: # e.g. 110kV, adjust if needed
                     r_per_km, x_per_km, b_per_km, s_nom_typical = 0.15, 0.4, 2.0e-6, 250

                merged_network.add(
                    "Line",
                    name=new_line_name,
                    bus0=bus0_final_name,
                    bus1=bus1_final_name,
                    r=r_per_km * length_km,
                    x=x_per_km * length_km,
                    b=b_per_km * length_km,
                    s_nom=s_nom_typical, # Or s_nom based on more detailed line type data
                    carrier="AC",
                    type=f"estimated_{int(v_nom)}kV"
                )
        print(f"Added {len(new_line_candidates)} new lines to connect unmerged networkAT buses.")

    # 8. SUMMARY
    print("\n--- Merging Summary ---")
    print(f"Initial buses in network1: {len(network1.buses)}")
    print(f"Initial Austrian buses in network1: {len(austrian_buses1_df) if not austrian_buses1_df.empty else 0}")
    print(f"Reference buses selected from networkAT: {len(reference_buses2_df)}")
    print(f"Coordinate updates applied to existing network1 buses: {len(updates_for_network1_coords)}")
    print(f"Unmerged networkAT reference buses added as new: {len(unmerged_buses_from_networkAT_df)}")
    print(f"New lines added to connect these new buses: {len(new_line_candidates)}")
    print(f"Total buses in merged_network: {len(merged_network.buses)}")
    print(f"Total lines in merged_network: {len(merged_network.lines)}")
    print("--- Process Completed ---")

    return merged_network

if __name__ == '__main__':
    # --- Create Dummy PyPSA Network Objects for Demonstration ---
    # This section is for testing. In a real scenario, you would load your networks.

    # Dummy Network 1 (Europe-wide, simplified)
    net1 = pypsa.Network()
    net1.add("Bus", "EU_Bus_DE1", v_nom=380, x=10.0, y=50.0, country="DE")
    net1.add("Bus", "EU_Bus_AT1", v_nom=380, x=13.0, y=48.0, country="AT") # Austrian bus to be updated
    net1.add("Bus", "EU_Bus_AT2", v_nom=220, x=14.0, y=47.5, country="AT") # Another Austrian bus
    net1.add("Bus", "EU_Bus_FR1", v_nom=380, x=2.0, y=46.0, country="FR")
    net1.add("Line", "L_DE_AT", bus0="EU_Bus_DE1", bus1="EU_Bus_AT1", r=0.01, x=0.1, s_nom=1000)

    # Dummy Network 2 (Detailed Austria)
    net2 = pypsa.Network()
    # This bus should match EU_Bus_AT1 and update its coordinates
    net2.add("Bus", "AT_Sub_West", v_nom=380, x=13.05, y=48.05, country="AT", type="3") # type as string for pypsa
    # This bus is new and should be added, then connected
    net2.add("Bus", "AT_Sub_South_Border", v_nom=220, x=14.5, y=46.5, country="AT", type="3")
    # Another reference bus in AT to connect AT_Sub_South_Border to
    net2.add("Bus", "AT_Sub_Central", v_nom=220, x=14.2, y=47.0, country="AT", type="3")
    net2.add("Bus", "AT_Sub_NonRef", v_nom=110, x=13.5, y=47.8, country="AT", type="1") # Not a reference bus

    print("Dummy network1 buses:\n", net1.buses)
    print("Dummy networkAT buses:\n", net2.buses)

    # Perform the merge
    # Assuming coordinates are projected for simplicity in dummy data, native units (no conversion to km needed for distance logic itself)
    merged_network_result = merge_european_and_austrian_networks(net1, net2,
                                                                 coords_are_lat_lon=False,
                                                                 convert_distance_to_km=False)

    print("\nMerged network buses:\n", merged_network_result.buses)
    print("\nMerged network lines:\n", merged_network_result.lines)

    # Example of checking a specific bus coordinate update:
    print("\nCoordinates of EU_Bus_AT1 in original net1:", net1.buses.loc["EU_Bus_AT1", ["x", "y"]].to_dict())
    print("Coordinates of EU_Bus_AT1 in merged_network:", merged_network_result.buses.loc["EU_Bus_AT1", ["x", "y"]].to_dict())

    # Check newly added bus and line
    if "AT_NET2_AT_Sub_South_Border" in merged_network_result.buses.index:
        print("\nNewly added bus AT_NET2_AT_Sub_South_Border:\n", merged_network_result.buses.loc["AT_NET2_AT_Sub_South_Border"])
        # Find lines connected to this new bus
        connected_lines = merged_network_result.lines[
            (merged_network_result.lines.bus0 == "AT_NET2_AT_Sub_South_Border") |
            (merged_network_result.lines.bus1 == "AT_NET2_AT_Sub_South_Border")
        ]
        print("\nLines connected to AT_NET2_AT_Sub_South_Border:\n", connected_lines)
```

```python
import numpy as np
import geopandas as gpd
from scipy.spatial import cKDTree
from shapely.geometry import Point

def match_and_update_border_buses(network1, networkAT, voltage_min=220.0, voltage_max=380.0, border_type="3", max_dist_km=50):
    # 1. Get reference border buses from networkAT
    border_buses2 = networkAT.buses[
        (networkAT.buses['type'] == border_type) &
        (networkAT.buses['v_nom'] >= voltage_min) &
        (networkAT.buses['v_nom'] <= voltage_max)
    ]
    
    if border_buses2.empty:
        print("No border buses found in networkAT.")
        return network1, [], []

    # 2. Create GeoDataFrames for spatial operations
    gdf1 = gpd.GeoDataFrame(network1.buses.copy(), geometry=gpd.points_from_xy(network1.buses.x, network1.buses.y), crs="EPSG:4326").to_crs(3857)
    gdf2 = gpd.GeoDataFrame(border_buses2.copy(), geometry=gpd.points_from_xy(border_buses2.x, border_buses2.y), crs="EPSG:4326").to_crs(3857)

    coords1 = np.array([(pt.x, pt.y) for pt in gdf1.geometry])
    coords2 = np.array([(pt.x, pt.y) for pt in gdf2.geometry])

    tree = cKDTree(coords1)
    distances, indices = tree.query(coords2, k=1)
    distances_km = distances / 1000

    # 3. Apply matches
    matched_buses = {}
    unmatched_border_buses = []
    
    for i, (dist_km, idx1) in enumerate(zip(distances_km, indices)):
        if dist_km <= max_dist_km:
            ref_bus2 = border_buses2.iloc[i]
            target_bus1 = gdf1.iloc[idx1]

            # Update network1 bus coordinates to match networkAT's bus
            network1.buses.at[target_bus1.name, 'x'] = ref_bus2['x']
            network1.buses.at[target_bus1.name, 'y'] = ref_bus2['y']

            matched_buses[ref_bus2.name] = target_bus1.name
        else:
            unmatched_border_buses.append(border_buses2.index[i])

    used_network1_buses = list(matched_buses.values())
    unused_networkAT_buses = list(border_buses2.index.difference(matched_buses.keys()))

    return network1, unmatched_border_buses, unused_networkAT_buses
```

```python
def remove_aggressive_redundant_lines(network, only_border_lines=False, keep='max_s_nom'):
    """
    Remove redundant lines (same unordered bus pairs).
    
    Parameters:
        network: pypsa.Network
        only_border_lines: bool - only consider lines starting with "border_"
        keep: "first" | "last" | "max_s_nom" - which line to keep among duplicates
    """
    from collections import defaultdict

    pair_to_lines = defaultdict(list)

    for line_name, line in network.lines.iterrows():
        if only_border_lines and not line_name.startswith("border_"):
            continue
        bus_pair = tuple(sorted((line.bus0, line.bus1)))
        pair_to_lines[bus_pair].append((line_name, line))

    to_drop = []

    for bus_pair, lines in pair_to_lines.items():
        if len(lines) <= 1:
            continue

        # Decide which line to keep
        if keep == "max_s_nom":
            lines_sorted = sorted(lines, key=lambda x: x[1].s_nom if pd.notnull(x[1].s_nom) else 0, reverse=True)
        elif keep == "last":
            lines_sorted = lines[::-1]
        else:  # "first"
            lines_sorted = lines

        # Keep only the first one
        for line_name, _ in lines_sorted[1:]:
            print(f"🗑️ Removing redundant line: {line_name} between {bus_pair[0]} ↔ {bus_pair[1]}")
            to_drop.append(line_name)

    network.lines.drop(index=to_drop, inplace=True)
    print(f"✅ Removed {len(to_drop)} redundant line(s).")
```

```python
from math import radians, cos, sin, sqrt, atan2

def haversine_km(x1, y1, x2, y2):
    # Earth radius in km
    R = 6371.0
    dlon = radians(x2 - x1)
    dlat = radians(y2 - y1)
    a = sin(dlat / 2)**2 + cos(radians(y1)) * cos(radians(y2)) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c
```

```python
from shapely.geometry import Point, LineString
from scipy.spatial import cKDTree
import numpy as np

def haversine_km(lon1, lat1, lon2, lat2):
    # Calculate great-circle distance in km between two points (lon, lat) in degrees
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a)) 
    r = 6371  # Earth radius in km
    return c * r

def connect_or_merge_buses(network, unmatched, max_line_dist_km=10, fallback_dist_km=100, line_s_nom=100):
    coords = network.buses[["x", "y"]]
    austria_mask = network.buses.index.str.startswith("AT")
    
    # Build cKDTree for fast nearest neighbor queries (in lon/lat)
    austria_ids = coords[austria_mask].index.to_numpy()
    non_austria_ids = coords[~austria_mask].index.to_numpy()
    austria_tree = cKDTree(coords.loc[austria_ids].values)
    non_austria_tree = cKDTree(coords.loc[non_austria_ids].values)
    
    for bus in unmatched:
        if bus not in network.buses.index:
            print(f"⚠️ Bus {bus} not in network.")
            continue

        xb, yb = coords.loc[bus]  # lon, lat
        bus_point = Point(xb, yb)
        inserted = False
        
        # Step 1: Try to find a nearby line to insert into
        for line in network.lines.itertuples():
            x0, y0 = coords.loc[line.bus0]
            x1, y1 = coords.loc[line.bus1]

            line_geom = LineString([(x0, y0), (x1, y1)])
            nearest_point = line_geom.interpolate(line_geom.project(Point(xb, yb)))
            x_near, y_near = nearest_point.x, nearest_point.y
            
            dist = haversine_km(xb, yb, x_near, y_near)  # km

            if dist <= max_line_dist_km:
                print(f"🔀 Inserting '{bus}' into line {line.Index} between {line.bus0} and {line.bus1} (dist={dist:.2f} km)")
                
                # Remove original line
                network.remove("Line", line.Index)
                
                # Calculate lengths of new segments using haversine
                d0 = haversine_km(x0, y0, xb, yb)
                d1 = haversine_km(x1, y1, xb, yb)

                # Add two new line segments replacing the original line
                network.add("Line", name=f"{line.Index}_split1",
                            bus0=line.bus0, bus1=bus,
                            r=0.01, x=0.05, s_nom=line.s_nom, length=d0)
                network.add("Line", name=f"{line.Index}_split2",
                            bus0=bus, bus1=line.bus1,
                            r=0.01, x=0.05, s_nom=line.s_nom, length=d1)

                inserted = True
                break
        
        if inserted:
            continue  # done with this bus

        # Step 2: Fallback connect to 1 Austrian + 1 foreign bus
        pos = np.array([[xb, yb]])
        
        dist_in, idx_in = austria_tree.query(pos, k=1, distance_upper_bound=fallback_dist_km)
        dist_in = dist_in[0]
        idx_in = idx_in[0]
        if idx_in < len(austria_ids):
            target = austria_ids[idx_in]
            d_km = haversine_km(xb, yb, *coords.loc[target])
            print(f"🔁 Fallback: '{bus}' → Austria: '{target}' ({d_km:.2f} km)")
            network.add("Line", name=f"fallback_in_{bus}_{target}",
                        bus0=bus, bus1=target,
                        r=0.01, x=0.05, s_nom=line_s_nom,
                        length=d_km)
        else:
            print(f"⚠️ No Austrian fallback neighbor found for '{bus}' within {fallback_dist_km} km")

        dist_out, idx_out = non_austria_tree.query(pos, k=1, distance_upper_bound=fallback_dist_km)
        dist_out = dist_out[0]
        idx_out = idx_out[0]
        if idx_out < len(non_austria_ids):
            target = non_austria_ids[idx_out]
            d_km = haversine_km(xb, yb, *coords.loc[target])
            print(f"🔁 Fallback: '{bus}' → Outside: '{target}' ({d_km:.2f} km)")
            network.add("Line", name=f"fallback_out_{bus}_{target}",
                        bus0=bus, bus1=target,
                        r=0.01, x=0.05, s_nom=line_s_nom,
                        length=d_km)
        else:
            print(f"⚠️ No foreign fallback neighbor found for '{bus}' within {fallback_dist_km} km")

    print("✅ Finished connecting/merging unmatched buses.")
```

```python
import geopandas as gpd
from shapely.geometry import Point, LineString
from scipy.spatial import cKDTree
import numpy as np

def haversine_km(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    dlon, dlat = lon2 - lon1, lat2 - lat1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    return 6371 * 2 * np.arcsin(np.sqrt(a))  # Earth radius in km

# def assign_countries(coords, country_shapes):
#     points = gpd.GeoSeries([Point(xy) for xy in coords[["x", "y"]].to_numpy()], crs="EPSG:4326")
#     coords_gdf = gpd.GeoDataFrame(coords, geometry=points)
#     coords_gdf = gpd.sjoin(coords_gdf, country_shapes[["geometry", "ISO_A2"]], how="left", predicate="within")
#     coords["country"] = coords_gdf["ISO_A2"]
#     return coords

# def assign_countries(coords):
#     # Assume bus names start with country code (like "AT", "DE", etc.)
#     coords["country"] = coords.index.str.extract(r"^([A-Z]{2})")[0]
#     return coords

# def assign_countries(coords, country_shapes):
#     # Convert x/y into geometries
#     coords = coords.copy()
#     coords["geometry"] = [Point(xy) for xy in coords[["x", "y"]].to_numpy()]
#     coords_gdf = gpd.GeoDataFrame(coords, geometry="geometry", crs="EPSG:4326")

#     # Spatial join: find which country each point lies within
#     joined = gpd.sjoin(coords_gdf, country_shapes[["geometry", "country"]], how="left", predicate="within")
#     coords["country"] = joined["country"].values
    
#     return coords

# def assign_countries(buses_df, country_shapes):
#     """
#     Assigns countries to buses based on a three-stage process:
#     1. Extracts country code from the bus index (e.g., "CH0 xxx").
#     2. If not found, extracts country code from 'substation_name' (e.g., "Facility (XX)").
#     3. If still not found, uses spatial join with country_shapes.

#     Args:
#         buses_df (pd.DataFrame): DataFrame of buses. Must include:
#                                  - Index (bus names/IDs).
#                                  - 'x' and 'y' columns for coordinates.
#                                  - Optionally, a 'substation_name' column.
#         country_shapes (gpd.GeoDataFrame): GeoDataFrame with country polygons.
#                                           Must have a geometry column and a column
#                                           indicating the country code (e.g., 'country' or 'ISO_A2').

#     Returns:
#         pd.DataFrame: The input buses_df with an added/updated 'country' column.
#     """
#     # Use a temporary column for assignments to allow sequential filling
#     buses_df['country_assigned'] = pd.NA

#     # --- Stage 1: Extract from bus index ---
#     # Regex to find two uppercase letters at the beginning of the index string
#     idx_country_codes = buses_df.index.str.extract(r"^([A-Z]{2})", expand=False)
#     buses_df['country_assigned'] = buses_df['country_assigned'].fillna(idx_country_codes)

#     # --- Stage 2: Extract from substation_name for remaining NaNs ---
#     if 'substation_name' in buses_df.columns:
#         # Regex to find two uppercase letters within parentheses in substation_name
#         # Ensure substation_name is string type to use .str accessor; NaNs become "nan"
#         sub_country_codes = buses_df['substation_name'].astype(str).str.extract(r"\(([A-Z]{2})\)", expand=False)
#         buses_df['country_assigned'] = buses_df['country_assigned'].fillna(sub_country_codes)
    
#     # --- Stage 3: Spatial join for remaining NaNs ---
#     mask_remaining_na = buses_df['country_assigned'].isna()
#     if mask_remaining_na.any():
#         # Get the subset of buses that still need country assignment
#         # Work on a copy to avoid SettingWithCopyWarning when adding 'geometry'
#         buses_to_sjoin = buses_df[mask_remaining_na].copy()

#         # Ensure 'x' and 'y' columns are numeric and valid
#         buses_to_sjoin['x'] = pd.to_numeric(buses_to_sjoin['x'], errors='coerce')
#         buses_to_sjoin['y'] = pd.to_numeric(buses_to_sjoin['y'], errors='coerce')
        
#         # Drop rows where x or y became NaN after coercion or were already NaN
#         buses_to_sjoin.dropna(subset=['x', 'y'], inplace=True)
        
#         if not buses_to_sjoin.empty:
#             try:
#                 # Create GeoDataFrame for spatial join
#                 geometry = [Point(xy) for xy in buses_to_sjoin[["x", "y"]].to_numpy()]
#                 # Temporarily suppress UserWarning about CRS not being set for older geopandas versions
#                 # For newer versions, this is handled better or needs explicit crs_transform.
#                 # Assuming input x,y are EPSG:4326
#                 coords_gdf = gpd.GeoDataFrame(buses_to_sjoin, geometry=geometry, crs="EPSG:4326")

#                 # Prepare country_shapes: ensure it has a 'country' column for the join
#                 country_col_name_in_shapes = None
#                 if "country" in country_shapes.columns:
#                     country_col_name_in_shapes = "country"
#                 elif "ISO_A2" in country_shapes.columns:
#                     country_col_name_in_shapes = "ISO_A2"
#                 # Add more potential column names if needed, e.g. "ADM0_A3" etc.
                
#                 if country_col_name_in_shapes:
#                     country_shapes_for_join = country_shapes[["geometry", country_col_name_in_shapes]].rename(
#                         columns={country_col_name_in_shapes: "country_geom"} # Use a distinct name
#                     )
#                 else:
#                     # Fallback or error if no suitable country identifier column is found
#                     print("⚠️ Warning: 'country_shapes' GeoDataFrame does not have a recognized country identifier column (e.g., 'country', 'ISO_A2'). Spatial join may not provide country names.")
#                     # Attempt join anyway; it might fail or user might have a custom column
#                     country_shapes_for_join = country_shapes 


#                 # Perform spatial join
#                 # joined's index will be from coords_gdf (left DataFrame)
#                 joined = gpd.sjoin(coords_gdf, country_shapes_for_join, how="left", predicate="within")
                
#                 # The country information from country_shapes will be in 'country_geom' (or original name if rename didn't happen)
#                 # We need to handle the case where 'country_geom' might not be in `joined.columns` if the setup failed.
#                 if 'country_geom' in joined.columns:
#                     spatially_assigned_countries = joined['country_geom'].dropna()
#                     # Update the 'country_assigned' column in the original buses_df
#                     if not spatially_assigned_countries.empty:
#                         buses_df['country_assigned'].update(spatially_assigned_countries)
#                 elif country_col_name_in_shapes in joined.columns and country_col_name_in_shapes != "country_geom": # If rename didn't apply but original col name exists
#                     spatially_assigned_countries = joined[country_col_name_in_shapes].dropna()
#                     if not spatially_assigned_countries.empty:
#                         buses_df['country_assigned'].update(spatially_assigned_countries)


#             except Exception as e:
#                 print(f"⚠️ Error during spatial join: {e}. Some countries may not be assigned spatially.")

#     # Finalize: copy the assignments to the 'country' column and drop the temporary one
#     buses_df['country'] = buses_df['country_assigned']
#     del buses_df['country_assigned']
    
#     return buses_df

import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
# numpy is also used implicitly by pandas/geopandas

# (Keep your other functions like haversine_km, etc.)

def assign_countries(buses_df, country_shapes):
    """
    Assigns countries to buses based on a three-stage process:
    1. Extracts country code from the bus index (e.g., "CH0 xxx").
    2. If not found, extracts country code from 'substation_name' (e.g., "Facility (XX)").
    3. If still not found, uses spatial join with country_shapes.

    Args:
        buses_df (pd.DataFrame): DataFrame of buses. Must include:
                                 - Index (bus names/IDs).
                                 - 'x' and 'y' columns for coordinates.
                                 - Optionally, a 'substation_name' column.
        country_shapes (gpd.GeoDataFrame): GeoDataFrame with country polygons.
                                          Must have a geometry column and a column
                                          indicating the country code (e.g., 'country' or 'ISO_A2').

    Returns:
        pd.DataFrame: The input buses_df with an added/updated 'country' column.
    """

    # --- Stage 1: Extract from bus index ---
    # Regex to find two uppercase letters at the beginning of the index string.
    # This is expected to return a pandas Series.
    idx_country_codes = buses_df.index.str.extract(r"^([A-Z]{2})", expand=False)

    # Directly assign the extracted codes. This will create the 'country_assigned' column.
    # If idx_country_codes is an Index (which is unexpected but indicated by the error),
    # pandas will convert it to a Series upon assignment to a DataFrame column.
    # If it's already a Series (as expected), this assignment is also correct.
    buses_df['country_assigned'] = idx_country_codes

    # --- Stage 2: Extract from substation_name for remaining NaNs ---
    if 'substation_name' in buses_df.columns:
        sub_country_codes = buses_df['substation_name'].astype(str).str.extract(r"\(([A-Z]{2})\)", expand=False)
        # Now, 'country_assigned' is definitely a Series, so fillna will work as expected.
        buses_df['country_assigned'] = buses_df['country_assigned'].fillna(sub_country_codes)
    
    # --- Stage 3: Spatial join for remaining NaNs ---
    mask_remaining_na = buses_df['country_assigned'].isna()
    if mask_remaining_na.any():
        buses_to_sjoin = buses_df[mask_remaining_na].copy()

        buses_to_sjoin['x'] = pd.to_numeric(buses_to_sjoin['x'], errors='coerce')
        buses_to_sjoin['y'] = pd.to_numeric(buses_to_sjoin['y'], errors='coerce')
        buses_to_sjoin.dropna(subset=['x', 'y'], inplace=True)
        
        if not buses_to_sjoin.empty:
            try:
                geometry = [Point(xy) for xy in buses_to_sjoin[["x", "y"]].to_numpy()]
                coords_gdf = gpd.GeoDataFrame(buses_to_sjoin, geometry=geometry, crs="EPSG:4326")

                country_col_name_in_shapes = None
                if "country" in country_shapes.columns:
                    country_col_name_in_shapes = "country"
                elif "ISO_A2" in country_shapes.columns:
                    country_col_name_in_shapes = "ISO_A2"
                
                if country_col_name_in_shapes:
                    country_shapes_for_join = country_shapes[["geometry", country_col_name_in_shapes]].rename(
                        columns={country_col_name_in_shapes: "country_geom"}
                    )
                else:
                    print("⚠️ Warning: 'country_shapes' GeoDataFrame does not have a recognized country identifier column (e.g., 'country', 'ISO_A2'). Spatial join may not provide country names.")
                    country_shapes_for_join = country_shapes 

                joined = gpd.sjoin(coords_gdf, country_shapes_for_join, how="left", predicate="within")
                
                if 'country_geom' in joined.columns:
                    spatially_assigned_countries = joined['country_geom'].dropna()
                    if not spatially_assigned_countries.empty:
                        buses_df['country_assigned'].update(spatially_assigned_countries)
                elif country_col_name_in_shapes in joined.columns and country_col_name_in_shapes != "country_geom":
                    spatially_assigned_countries = joined[country_col_name_in_shapes].dropna()
                    if not spatially_assigned_countries.empty:
                        buses_df['country_assigned'].update(spatially_assigned_countries)
            except Exception as e:
                print(f"⚠️ Error during spatial join: {e}. Some countries may not be assigned spatially.")

    buses_df['country'] = buses_df['country_assigned']
    del buses_df['country_assigned']
    
    return buses_df

# The connect_or_merge_buses function would remain as previously defined,
# as the change is internal to assign_countries.

# def connect_or_merge_buses(network, unmatched, country_shapes, max_line_dist_km=10, fallback_dist_km=100, line_s_nom=100):
#     coords = network.buses[["x", "y"]].copy()
#     # coords = assign_countries(coords, country_shapes)
#     coords = assign_countries(coords, country_shapes)


#     # Build KD-trees per country
#     tree_by_country = {
#         country: cKDTree(group[["x", "y"]].values)
#         for country, group in coords.groupby("country")
#         if country is not np.nan
#     }

#     for bus in unmatched:
#         if bus not in network.buses.index:
#             print(f"⚠️ Bus {bus} not in network.")
#             continue

#         xb, yb = coords.loc[bus, ["x", "y"]]
#         bus_point = Point(xb, yb)
#         inserted = False

#         # Step 1: Try to insert into a nearby line
#         for line in network.lines.itertuples():
#             x0, y0 = coords.loc[line.bus0, ["x", "y"]]
#             x1, y1 = coords.loc[line.bus1, ["x", "y"]]

#             line_geom = LineString([(x0, y0), (x1, y1)])
#             nearest_point = line_geom.interpolate(line_geom.project(bus_point))
#             x_near, y_near = nearest_point.x, nearest_point.y
#             dist = haversine_km(xb, yb, x_near, y_near)

#             if dist <= max_line_dist_km:
#                 print(f"🔀 Inserting '{bus}' into line {line.Index} between {line.bus0} and {line.bus1} (dist={dist:.2f} km)")

#                 network.remove("Line", line.Index)

#                 d0 = haversine_km(x0, y0, xb, yb)
#                 d1 = haversine_km(x1, y1, xb, yb)

#                 network.add("Line", name=f"{line.Index}_split1",
#                             bus0=line.bus0, bus1=bus,
#                             r=0.01, x=0.05, s_nom=line.s_nom, length=d0)
#                 network.add("Line", name=f"{line.Index}_split2",
#                             bus0=bus, bus1=line.bus1,
#                             r=0.01, x=0.05, s_nom=line.s_nom, length=d1)

#                 inserted = True
#                 break

#         if inserted:
#             continue  # done with this bus

#         # Step 2: Fallback connections
#         bus_country = coords.loc[bus, "country"]
#         if pd.isna(bus_country):
#             print(f"⚠️ Could not detect country for {bus}. Skipping fallback.")
#             continue

#         def find_nearest(tree, ids, label):
#             pos = np.array([[xb, yb]])
#             dist, idx = tree.query(pos, k=1, distance_upper_bound=fallback_dist_km)
#             dist = dist[0]
#             idx = idx[0]
#             if idx < len(ids):
#                 target = ids[idx]
#                 d_km = haversine_km(xb, yb, *coords.loc[target, ["x", "y"]])
#                 print(f"🔁 Fallback: '{bus}' → {label}: '{target}' ({d_km:.2f} km)")
#                 network.add("Line", name=f"fallback_{label}_{bus}_{target}",
#                             bus0=bus, bus1=target,
#                             r=0.01, x=0.05, s_nom=line_s_nom,
#                             length=d_km)
#                 return True
#             else:
#                 print(f"⚠️ No {label} fallback found for '{bus}' within {fallback_dist_km} km")
#                 return False

#         # Austrian connection
#         # if "AT" in tree_by_country:
#         #     at_ids = coords[coords["country"] == "AT"].index.to_numpy()
#         #     find_nearest(tree_by_country["AT"], at_ids, "austria")
#         # Define for AT or per country
#         neighbor_countries = {
#             "AT": ["DE", "IT", "CH", "CZ", "SI", "HU", "SK"],
#             "DE": ["AT", "CH", "PL", "NL", "BE", "DK"],
#             # etc., extend as needed
#         }

#         # Try fallback to neighbors
#         for neighbor in neighbor_countries.get(bus_country, []):
#             if neighbor in tree_by_country:
#                 neighbor_ids = coords[coords["country"] == neighbor].index.to_numpy()
#                 if find_nearest(tree_by_country[neighbor], neighbor_ids, f"{bus_country}-{neighbor}"):
#                     break  # stop after first successful connection

#         # Same-country connection
#         if bus_country in tree_by_country and bus_country != "AT":
#             ids = coords[coords["country"] == bus_country].index.to_numpy()
#             find_nearest(tree_by_country[bus_country], ids, bus_country)

#     print("✅ Finished connecting/merging unmatched buses.")

# Your existing imports
import geopandas as gpd
from shapely.geometry import Point, LineString
from scipy.spatial import cKDTree
import numpy as np
import pandas as pd # Ensure pandas is imported

# ... (Haversine function, new assign_countries function from above) ...

def haversine_km(lon1, lat1, lon2, lat2): # Copied from your snippet for completeness
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    dlon, dlat = lon2 - lon1, lat2 - lat1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    return 6371 * 2 * np.arcsin(np.sqrt(a))  # Earth radius in km

def connect_or_merge_buses(network, unmatched, country_shapes, max_line_dist_km=10, fallback_dist_km=100, line_s_nom=100):
    # --- MODIFIED SECTION ---
    # It's best to work with a copy of the buses DataFrame to avoid modifying the original network object
    # directly within assign_countries, unless that's the specific design.
    # This copy will include all necessary columns like index, 'x', 'y', and 'substation_name'.
    buses_data_copy = network.buses.copy()

    # Call the new assign_countries function.
    # This function will add/update the 'country' column in the `buses_data_copy` DataFrame.
    buses_with_countries = assign_countries(buses_data_copy, country_shapes)
    
    # The rest of the function can use 'buses_with_countries' as 'coords'.
    # It contains all original bus data plus the accurately assigned 'country' column.
    coords = buses_with_countries
    # --- END OF MODIFIED SECTION ---

    # Build KD-trees per country
    # Ensure to handle cases where 'country' might still be NaN after assignment attempts
    tree_by_country = {
        country: cKDTree(group[["x", "y"]].values)
        for country, group in coords.groupby("country")
        if pd.notna(country) and not group[["x", "y"]].dropna().empty # Check for valid country and non-empty valid coordinates
    }

    for bus in unmatched:
        if bus not in network.buses.index: # Check against original network.buses index
            print(f"⚠️ Bus {bus} not in network.")
            continue

        # Use `coords` (which is buses_with_countries) to get bus details
        if bus not in coords.index: # Should not happen if unmatched is from network.buses
            print(f"⚠️ Bus {bus} not in processed coords DataFrame. Skipping.")
            continue

        # Ensure the bus has valid coordinate data in `coords`
        if coords.loc[bus, ["x", "y"]].isnull().any():
            print(f"⚠️ Bus {bus} has missing coordinate data. Skipping.")
            continue
            
        xb, yb = coords.loc[bus, ["x", "y"]]
        bus_point = Point(xb, yb)
        inserted = False

        # Step 1: Try to insert into a nearby line
        for line_idx, line in network.lines.iterrows(): # Use iterrows() for DataFrame
            # Ensure bus0 and bus1 of the line exist in coords and have valid coordinates
            if not (line.bus0 in coords.index and line.bus1 in coords.index and \
                    coords.loc[line.bus0, ["x", "y"]].notnull().all() and \
                    coords.loc[line.bus1, ["x", "y"]].notnull().all()):
                print(f"⚠️ Line {line_idx} has buses ({line.bus0}, {line.bus1}) with missing data in coords. Skipping this line for bus {bus}.")
                continue

            x0, y0 = coords.loc[line.bus0, ["x", "y"]]
            x1, y1 = coords.loc[line.bus1, ["x", "y"]]

            line_geom = LineString([(x0, y0), (x1, y1)])
            # Handle cases where line_geom might be invalid (e.g. zero length)
            if line_geom.length == 0:
                continue 
            nearest_point = line_geom.interpolate(line_geom.project(bus_point))
            x_near, y_near = nearest_point.x, nearest_point.y
            dist = haversine_km(xb, yb, x_near, y_near)

            if dist <= max_line_dist_km:
                line_name = line.name if pd.notna(line.name) else line_idx # Handle unnamed lines
                print(f"🔀 Inserting '{bus}' into line {line_name} between {line.bus0} and {line.bus1} (dist={dist:.2f} km)")

                network.remove("Line", line_idx) # Use line_idx (original index from iterrows)

                d0 = haversine_km(x0, y0, xb, yb)
                d1 = haversine_km(x1, y1, xb, yb)

                network.add("Line", name=f"{line_name}_split1",
                            bus0=line.bus0, bus1=bus,
                            r=0.01, x=0.05, s_nom=line.s_nom, length=d0) # Ensure line.s_nom is valid
                network.add("Line", name=f"{line_name}_split2",
                            bus0=bus, bus1=line.bus1,
                            r=0.01, x=0.05, s_nom=line.s_nom, length=d1)

                inserted = True
                break
        
        if inserted:
            continue

        # Step 2: Fallback connections
        bus_country = coords.loc[bus, "country"] # Get country from the processed coords
        if pd.isna(bus_country):
            print(f"⚠️ Could not detect country for {bus}. Skipping fallback connections.")
            continue

        # (The find_nearest function and fallback logic seems mostly okay but ensure
        #  it correctly uses `coords` which now is the full `buses_with_countries` DataFrame.
        #  The `ids` are indices of the `coords` DataFrame.)
        # Make sure find_nearest is defined or part of this scope.
        # Assuming find_nearest is defined similar to your original structure:
        def find_nearest(tree, target_ids_in_coords, label_suffix):
            pos = np.array([[xb, yb]]) # xb, yb of the current `bus`
            dist_query, idx_query = tree.query(pos, k=1, distance_upper_bound=fallback_dist_km / 6371 * (180/np.pi) ) # Approx conversion for KDTree if it uses radians/degrees
            # Note: cKDTree uses Euclidean distance on the provided coordinates.
            # If x,y are lon/lat, Euclidean distance isn't true geographic distance.
            # For small fallback_dist_km, it might be an acceptable approximation if coordinates are locally pseudo-Cartesian.
            # Otherwise, a more geographically aware nearest neighbor search is needed or filter post-query.
            # The original code queries with fallback_dist_km which is likely intended for Haversine.
            # The cKDTree distance_upper_bound is in the same units as the tree data (x,y).
            # A simple fix is to query more candidates (e.g. k=5 or 10) and then compute Haversine for them.
            # For now, let's assume current KDTree usage is what you intend.

            dist_query = dist_query[0]
            idx_query = idx_query[0]

            if idx_query < len(target_ids_in_coords): # Check if a point was found within the bound
                target_bus_id = target_ids_in_coords[idx_query]
                # Re-calculate actual Haversine distance
                d_km = haversine_km(xb, yb, *coords.loc[target_bus_id, ["x", "y"]])

                if d_km <= fallback_dist_km : # Check Haversine distance against the threshold
                    print(f"🔁 Fallback: '{bus}' ({bus_country}) → {label_suffix}: '{target_bus_id}' ({coords.loc[target_bus_id, 'country']}) ({d_km:.2f} km)")
                    network.add("Line", name=f"fallback_{label_suffix}_{bus}_{target_bus_id}",
                                bus0=bus, bus1=target_bus_id,
                                r=0.01, x=0.05, s_nom=line_s_nom, # Ensure line_s_nom is defined
                                length=d_km)
                    return True
                else:
                    print(f"⚠️ Candidate {target_bus_id} from KDTree for '{bus}' is too far by Haversine ({d_km:.2f} km > {fallback_dist_km} km).")
                    return False # Candidate found by KDTree but failed Haversine check
            else:
                # This means no point found by KDTree within its distance_upper_bound
                print(f"⚠️ No {label_suffix} fallback candidate found by KDTree for '{bus}' within its search radius.")
                return False


        # Define neighbor_countries (as in your example)
        neighbor_countries = {
            "AT": ["DE", "IT", "CH", "CZ", "SI", "HU", "SK"],
            "DE": ["AT", "CH", "PL", "NL", "BE", "DK", "CZ", "FR", "LU"], # Expanded DE
            "CH": ["AT", "DE", "IT", "FR", "LI"],
            "IT": ["AT", "CH", "FR", "SI"],
            # etc., extend as needed
        }
        
        connection_made = False
        # Try fallback to its own country first (if not already tried or if preferred)
        if bus_country in tree_by_country:
            country_ids = coords[coords["country"] == bus_country].index.to_numpy()
            # Exclude self from potential targets
            country_ids_no_self = country_ids[country_ids != bus]
            if len(country_ids_no_self) > 0:
                 if find_nearest(tree_by_country[bus_country], country_ids_no_self, f"{bus_country}-self"):
                    connection_made = True

        # If not connected within its own country, try neighbors
        if not connection_made:
            for neighbor in neighbor_countries.get(bus_country, []):
                if neighbor in tree_by_country:
                    neighbor_ids = coords[coords["country"] == neighbor].index.to_numpy()
                    if len(neighbor_ids) > 0: # Ensure there are buses in the neighbor country
                        if find_nearest(tree_by_country[neighbor], neighbor_ids, f"{bus_country}-{neighbor}"):
                            connection_made = True
                            break # Stop after first successful connection to a neighbor

        if not connection_made:
             print(f"⚠️ No fallback connection made for bus '{bus}' in country '{bus_country}' or its neighbors.")


    print("✅ Finished connecting/merging unmatched buses.")
```

```python
import geopandas as gpd
from shapely.geometry import Point, LineString
from scipy.spatial import cKDTree
import numpy as np
import pandas as pd

# def haversine_km(lon1, lat1, lon2, lat2):
#     """Calculate the great circle distance between two points on earth in kilometers"""
#     lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
#     dlon, dlat = lon2 - lon1, lat2 - lat1
#     a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
#     return 6371 * 2 * np.arcsin(np.sqrt(a))

def haversine_km(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    dlon, dlat = lon2 - lon1, lat2 - lat1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    return 6371 * 2 * np.arcsin(np.sqrt(a))  # Earth radius in km

def assign_countries(buses_df, country_shapes):
    """Assign countries to buses using multiple strategies"""
    # Stage 1: Extract from bus index
    idx_country_codes = buses_df.index.str.extract(r"^([A-Z]{2})", expand=False)
    buses_df['country_assigned'] = idx_country_codes

    # Stage 2: Extract from substation_name
    if 'substation_name' in buses_df.columns:
        sub_country_codes = buses_df['substation_name'].astype(str).str.extract(r"\(([A-Z]{2})\)", expand=False)
        buses_df['country_assigned'] = buses_df['country_assigned'].fillna(sub_country_codes)
    
    # Stage 3: Spatial join for remaining NaNs
    mask_remaining_na = buses_df['country_assigned'].isna()
    if mask_remaining_na.any():
        buses_to_sjoin = buses_df[mask_remaining_na].copy()
        buses_to_sjoin['x'] = pd.to_numeric(buses_to_sjoin['x'], errors='coerce')
        buses_to_sjoin['y'] = pd.to_numeric(buses_to_sjoin['y'], errors='coerce')
        buses_to_sjoin.dropna(subset=['x', 'y'], inplace=True)
        
        if not buses_to_sjoin.empty:
            try:
                geometry = [Point(xy) for xy in buses_to_sjoin[["x", "y"]].to_numpy()]
                coords_gdf = gpd.GeoDataFrame(buses_to_sjoin, geometry=geometry, crs="EPSG:4326")

                country_col_name_in_shapes = None
                if "country" in country_shapes.columns:
                    country_col_name_in_shapes = "country"
                elif "ISO_A2" in country_shapes.columns:
                    country_col_name_in_shapes = "ISO_A2"
                
                if country_col_name_in_shapes:
                    country_shapes_for_join = country_shapes[["geometry", country_col_name_in_shapes]].rename(
                        columns={country_col_name_in_shapes: "country_geom"}
                    )
                    joined = gpd.sjoin(coords_gdf, country_shapes_for_join, how="left", predicate="within")
                    
                    if 'country_geom' in joined.columns:
                        spatially_assigned_countries = joined['country_geom'].dropna()
                        if not spatially_assigned_countries.empty:
                            buses_df['country_assigned'].update(spatially_assigned_countries)
            except Exception as e:
                print(f"⚠️ Error during spatial join: {e}")

    buses_df['country'] = buses_df['country_assigned']
    del buses_df['country_assigned']
    return buses_df

def find_isolated_buses(network):
    """Find buses that have no connections (degree 0)"""
    # Get all buses that appear in lines
    connected_buses = set()
    for _, line in network.lines.iterrows():
        connected_buses.add(line.bus0)
        connected_buses.add(line.bus1)
    
    # Find buses not in any line
    all_buses = set(network.buses.index)
    isolated_buses = all_buses - connected_buses
    
    print(f"Found {len(isolated_buses)} isolated buses: {list(isolated_buses)}")
    return list(isolated_buses)

def connect_or_merge_buses(network, unmatched, country_shapes, max_line_dist_km=10, fallback_dist_km=100, line_s_nom=100):
    """Enhanced version with better isolated bus detection and connection strategies"""
    
    # First, let's find truly isolated buses
    isolated_buses = find_isolated_buses(network)
    
    # Combine provided unmatched with discovered isolated buses
    all_unmatched = list(set(unmatched + isolated_buses))
    print(f"Total buses to connect: {len(all_unmatched)}")
    
    # Process buses with country assignment
    buses_data_copy = network.buses.copy()
    buses_with_countries = assign_countries(buses_data_copy, country_shapes)
    coords = buses_with_countries

    # Build KD-trees per country for connected buses only
    connected_buses = set()
    for _, line in network.lines.iterrows():
        connected_buses.add(line.bus0)
        connected_buses.add(line.bus1)
    
    # Filter coords to only include connected buses for KD-tree construction
    connected_coords = coords[coords.index.isin(connected_buses)]
    
    tree_by_country = {}
    for country, group in connected_coords.groupby("country"):
        if pd.notna(country) and not group[["x", "y"]].dropna().empty:
            tree_by_country[country] = {
                'tree': cKDTree(group[["x", "y"]].values),
                'ids': group.index.to_numpy()
            }

    def find_nearest_connected_bus(bus_x, bus_y, target_country=None, max_dist_km=None):
        """Find nearest connected bus, optionally filtered by country"""
        best_dist = float('inf')
        best_bus = None
        
        countries_to_check = [target_country] if target_country else list(tree_by_country.keys())
        
        for country in countries_to_check:
            if country not in tree_by_country:
                continue
                
            tree_info = tree_by_country[country]
            tree = tree_info['tree']
            ids = tree_info['ids']
            
            # Query multiple nearest neighbors to handle edge cases
            k = min(10, len(ids))  # Get up to 10 nearest
            distances, indices = tree.query([[bus_x, bus_y]], k=k)
            
            # Handle both single result and multiple results
            if k == 1:
                # Single result case
                distances = [distances]
                indices = [indices]
            
            # Ensure we have arrays to iterate over
            distances = np.atleast_1d(distances.flatten()) if hasattr(distances, 'flatten') else np.atleast_1d(distances)
            indices = np.atleast_1d(indices.flatten()) if hasattr(indices, 'flatten') else np.atleast_1d(indices)
            
            for dist_approx, idx in zip(distances, indices):
                if idx < len(ids):
                    candidate_bus = ids[idx]
                    candidate_x, candidate_y = coords.loc[candidate_bus, ["x", "y"]]
                    actual_dist = haversine_km(bus_x, bus_y, candidate_x, candidate_y)
                    
                    if actual_dist < best_dist and (max_dist_km is None or actual_dist <= max_dist_km):
                        best_dist = actual_dist
                        best_bus = candidate_bus
        
        return best_bus, best_dist if best_bus else (None, float('inf'))

    neighbor_countries = {
        "AT": ["DE", "IT", "CH", "CZ", "SI", "HU", "SK"],
        "DE": ["AT", "CH", "PL", "NL", "BE", "DK", "CZ", "FR", "LU"],
        "CH": ["AT", "DE", "IT", "FR"],
        "IT": ["AT", "CH", "FR", "SI"],
        "CZ": ["AT", "DE", "SK", "PL"],
        "SI": ["AT", "IT", "HU"],
        "HU": ["AT", "SI", "SK"],
        "SK": ["AT", "CZ", "HU"],
    }

    for bus in all_unmatched:
        if bus not in network.buses.index:
            print(f"⚠️ Bus {bus} not in network.")
            continue

        if coords.loc[bus, ["x", "y"]].isnull().any():
            print(f"⚠️ Bus {bus} has missing coordinate data. Skipping.")
            continue
            
        xb, yb = coords.loc[bus, ["x", "y"]]
        bus_point = Point(xb, yb)
        inserted = False

        # Step 1: Try to insert into a nearby line (same as before)
        for line_idx, line in network.lines.iterrows():
            if not (line.bus0 in coords.index and line.bus1 in coords.index and 
                    coords.loc[line.bus0, ["x", "y"]].notnull().all() and 
                    coords.loc[line.bus1, ["x", "y"]].notnull().all()):
                continue

            x0, y0 = coords.loc[line.bus0, ["x", "y"]]
            x1, y1 = coords.loc[line.bus1, ["x", "y"]]

            line_geom = LineString([(x0, y0), (x1, y1)])
            if line_geom.length == 0:
                continue
                
            nearest_point = line_geom.interpolate(line_geom.project(bus_point))
            x_near, y_near = nearest_point.x, nearest_point.y
            dist = haversine_km(xb, yb, x_near, y_near)

            if dist <= max_line_dist_km:
                line_name = line.name if pd.notna(line.name) else line_idx
                print(f"🔀 Inserting '{bus}' into line {line_name} between {line.bus0} and {line.bus1} (dist={dist:.2f} km)")

                network.remove("Line", line_idx)

                d0 = haversine_km(x0, y0, xb, yb)
                d1 = haversine_km(x1, y1, xb, yb)

                network.add("Line", name=f"{line_name}_split1",
                            bus0=line.bus0, bus1=bus,
                            r=0.01, x=0.05, s_nom=line.s_nom, length=d0)
                network.add("Line", name=f"{line_name}_split2",
                            bus0=bus, bus1=line.bus1,
                            r=0.01, x=0.05, s_nom=line.s_nom, length=d1)

                inserted = True
                break

        if inserted:
            continue

        # Step 2: Enhanced fallback connections
        bus_country = coords.loc[bus, "country"]
        if pd.isna(bus_country):
            print(f"⚠️ Could not detect country for {bus}. Trying global connection...")
            # Try to connect to any nearby connected bus
            target_bus, dist = find_nearest_connected_bus(xb, yb, max_dist_km=fallback_dist_km)
            if target_bus:
                print(f"🔁 Global fallback: '{bus}' → '{target_bus}' ({dist:.2f} km)")
                network.add("Line", name=f"fallback_global_{bus}_{target_bus}",
                            bus0=bus, bus1=target_bus,
                            r=0.01, x=0.05, s_nom=line_s_nom, length=dist)
            continue

        connection_made = False
        
        # Strategy 1: Connect within same country
        if bus_country in tree_by_country:
            target_bus, dist = find_nearest_connected_bus(xb, yb, bus_country, fallback_dist_km)
            if target_bus and target_bus != bus:
                print(f"🔁 Same-country: '{bus}' ({bus_country}) → '{target_bus}' ({dist:.2f} km)")
                network.add("Line", name=f"fallback_{bus_country}_{bus}_{target_bus}",
                            bus0=bus, bus1=target_bus,
                            r=0.01, x=0.05, s_nom=line_s_nom, length=dist)
                connection_made = True

        # Strategy 2: Connect to neighboring countries
        if not connection_made:
            for neighbor in neighbor_countries.get(bus_country, []):
                if neighbor in tree_by_country:
                    target_bus, dist = find_nearest_connected_bus(xb, yb, neighbor, fallback_dist_km)
                    if target_bus:
                        print(f"🔁 Cross-border: '{bus}' ({bus_country}) → '{target_bus}' ({neighbor}) ({dist:.2f} km)")
                        network.add("Line", name=f"fallback_{bus_country}_{neighbor}_{bus}_{target_bus}",
                                    bus0=bus, bus1=target_bus,
                                    r=0.01, x=0.05, s_nom=line_s_nom, length=dist)
                        connection_made = True
                        break

        # Strategy 3: Last resort - connect to any nearby connected bus
        if not connection_made:
            target_bus, dist = find_nearest_connected_bus(xb, yb, max_dist_km=fallback_dist_km * 2)  # Double the search radius
            if target_bus:
                target_country = coords.loc[target_bus, "country"] if target_bus in coords.index else "Unknown"
                print(f"🔁 Last resort: '{bus}' ({bus_country}) → '{target_bus}' ({target_country}) ({dist:.2f} km)")
                network.add("Line", name=f"fallback_emergency_{bus}_{target_bus}",
                            bus0=bus, bus1=target_bus,
                            r=0.01, x=0.05, s_nom=line_s_nom, length=dist)
                connection_made = True

        if not connection_made:
            print(f"❌ Failed to connect bus '{bus}' ({bus_country})")

    print("✅ Finished connecting/merging unmatched buses.")

# Additional diagnostic function
def diagnose_network_connectivity(network):
    """Diagnose network connectivity issues"""
    print("\n=== Network Connectivity Diagnosis ===")
    
    # Count isolated buses
    isolated = find_isolated_buses(network)
    print(f"Isolated buses: {len(isolated)}")
    
    # Count buses by degree
    bus_degree = {}
    for bus in network.buses.index:
        bus_degree[bus] = 0
    
    for _, line in network.lines.iterrows():
        bus_degree[line.bus0] += 1
        bus_degree[line.bus1] += 1
    
    degree_counts = {}
    for degree in bus_degree.values():
        degree_counts[degree] = degree_counts.get(degree, 0) + 1
    
    print("Bus degree distribution:")
    for degree in sorted(degree_counts.keys()):
        print(f"  Degree {degree}: {degree_counts[degree]} buses")
    
    return isolated, bus_degree
```
