# merge_austrian_network
This notebook merges generated austrian network from postgis sources with the PyPSA-Eur generated european network with the aim to correctly display Austria's cross-border-lines.

## Workflow Overview
Austrian network is called networkAT
European network is called network1

1. Buses from network1 are compared with networkAT's buses and if gespatially differences exist, they shall be corrected 1:1 by networkAT buses coordinates. To find the approppriate buses a KDTree approach is used. Only cross-border nodes shall be considered, so the networkAT is filtered by type=3 and v_nom >= 110
2. There might be more cross-border nodes in networkAT than in network1, so not yet matched cross-border nodes are found and inserted into network1 as new nodes and connected to 2 neighboring nodes via new lines. Those 2 nodes shall be 1 in Austria and 1 in the corresponding foreign country.
3. If however there already is a parallel line existing, this line is cut and the cross-border node is inserted inbetween.

```python
# %% [markdown]
# # merge_austrian_network
# This notebook merges generated austrian network from postgis sources with the PyPSA-Eur generated european network with the aim to correctly display Austria's cross-border-lines.
# 
# ## Workflow Overview
# Austrian network is called networkAT
# European network is called network1
# 
# 1. Buses from network1 are compared with networkAT's buses and if gespatially differences exist, they shall be corrected 1:1 by networkAT buses coordinates. To find the approppriate buses a KDTree approach is used. Only cross-border nodes shall be considered, so the networkAT is filtered by type=3 and v_nom >= 110
# 2. There might be more cross-border nodes in networkAT than in network1, so not yet matched cross-border nodes are found and inserted into network1 as new nodes and connected to 2 neighboring nodes via new lines. Those 2 nodes shall be 1 in Austria and 1 in the corresponding foreign country.
# 3. If however there already is a parallel line existing, this line is cut and the cross-border node is inserted inbetween.
# 

# %% [markdown]
# ## 

# %% [markdown]
# ### imports

# %%
import pypsa
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
from scipy.spatial import cKDTree
import numpy as np
import os
```

```python

# %% [markdown]
# ### Load Networks

# %%
# Load the PyPSA-Europe and Austrian networks
# Create dummy snakemake inputs for standalone execution
class Snakemake:
    def __init__(self):
        # NOTE: You must replace these with the actual paths to your network files
        self.input = {
            "network1": "networks/elec.nc",
            "network2": "networks/austria.nc"
        }

try:
    snakemake
except NameError:
    snakemake = Snakemake()
    # Create dummy network files if they don't exist for demonstration
    if not os.path.exists("networks"):
        os.makedirs("networks")
    if not os.path.exists(snakemake.input.network1):
        dummy_net1 = pypsa.Network()
        dummy_net1.add("Bus", "DE Bus 1", x=10, y=50, v_nom=380, country="DE")
        dummy_net1.add("Bus", "IT Bus 1", x=12, y=46, v_nom=380, country="IT")
        dummy_net1.add("Bus", "AT Bus 1", x=14, y=48, v_nom=380, country="AT")
        dummy_net1.add("Line", "DE-AT", bus0="DE Bus 1", bus1="AT Bus 1", length=100, s_nom=1000, r=0.02, x=0.15)
        dummy_net1.export_to_netcdf(snakemake.input.network1)
    if not os.path.exists(snakemake.input.network2):
        dummy_net2 = pypsa.Network()
        dummy_net2.add("Bus", "AT CB Bus 1", x=13.5, y=47.5, v_nom=380, type="3") # type 3 is cross-border
        dummy_net2.export_to_netcdf(snakemake.input.network2)


network1 = pypsa.Network(snakemake.input.network1)
networkAT = pypsa.Network(snakemake.input.network2)

if network1.buses.empty or networkAT.buses.empty:
    raise ValueError("One of the networks is empty. Check the input files.")


print("Loaded network1 (Europe) with", len(network1.buses), "buses.")
print("Loaded networkAT (Austria) with", len(networkAT.buses), "buses.")

```

```python

# %% [markdown]
# ### Convert to GeoDataFrames

# %%
# Convert buses to GeoDataFrames (EPSG:4326)
buses1 = network1.buses.copy()
buses1["geometry"] = gpd.points_from_xy(buses1.x, buses1.y)
buses1_gdf = gpd.GeoDataFrame(buses1, geometry="geometry", crs="EPSG:4326")

busesAT = networkAT.buses.copy()
busesAT["geometry"] = gpd.points_from_xy(busesAT.x, busesAT.y)
busesAT_gdf = gpd.GeoDataFrame(busesAT, geometry="geometry", crs="EPSG:4326")


assert buses1_gdf.geometry.is_valid.all(), "Invalid geometries in network1"
assert busesAT_gdf.geometry.is_valid.all(), "Invalid geometries in networkAT"

```

```python

# %% [markdown]
# ### Clean Data & Filter Cross-Border Buses

# %%
# Clean and filter Austrian cross-border buses
# busesAT_gdf["type"] = (
#     busesAT_gdf["type"]
#     .astype(str)
#     .str.replace(".0", "", regex=False)
#     .astype(int)
# )
busesAT_gdf["type"] = pd.to_numeric(busesAT_gdf["type"], errors="coerce").astype("Int64")

# Remove all buses with index starting with 'AT'
busesAT_gdf = busesAT_gdf.loc[~busesAT_gdf.index.str.startswith("AT")]
busesAT_gdf["v_nom"] = pd.to_numeric(busesAT_gdf["v_nom"], errors="coerce")

# Then filter cross-border buses
cb_buses_AT = busesAT_gdf.query("type == 3 and v_nom >= 110").copy()

# Match substation_name with foreign country code pattern (e.g., "(D)", "(CZ)")
cb_buses_AT = cb_buses_AT[cb_buses_AT["substation_name"].str.contains(r"\([A-Z]{1,3}\)", na=False)]

# Extract the country code (e.g., "D", "CZ") from substation_name into a new column
cb_buses_AT["foreign_country"] = cb_buses_AT["substation_name"].str.extract(r"\(([A-Z]{1,3})\)")



print(f"Found {len(cb_buses_AT)} Austrian cross-border buses.")

```

```python

# %% [markdown]
# ### KDTree Matching with Haversine Distance

# %%
# Match Austrian CB buses to PyPSA-Eur using haversine KDTree

# network1_merged = network1.copy()

# def to_radians(df):
#     return np.radians(df[["y", "x"]].values)

# # Ensure there are cross-border buses to avoid errors
# if not cb_buses_AT.empty:
#     cb_coords_rad = to_radians(cb_buses_AT)
#     eu_coords_rad = to_radians(buses1_gdf)

#     tree = cKDTree(cb_coords_rad)
#     # Search within a 50km radius
#     distances_rad, indices = tree.query(eu_coords_rad, distance_upper_bound=50 / 6371.0)  

#     matched_count = 0
#     matched_ids = []

#     for i, (dist_rad, idx) in enumerate(zip(distances_rad, indices)):
#         # Check if a match was found within the radius
#         if np.isfinite(dist_rad):
#             bus1_id = buses1_gdf.index[i]
#             at_bus = cb_buses_AT.iloc[idx]
            
#             # Correct the coordinates in the merged network
#             network1_merged.buses.loc[bus1_id, ["x", "y"]] = at_bus.geometry.x, at_bus.geometry.y
#             matched_ids.append(cb_buses_AT.index[idx])
#             matched_count += 1

#     print(f"Matched and corrected coordinates for {matched_count} cross-border buses.")
# else:
#     matched_ids = []
#     print("No cross-border buses found in the Austrian network to match.")

from scipy.spatial import cKDTree
import numpy as np

def to_radians(df):
    return np.radians(df[["y", "x"]].values)

network1_merged = network1.copy()

if not cb_buses_AT.empty:
    cb_coords_rad = to_radians(cb_buses_AT)
    eu_coords_rad = to_radians(buses1_gdf)

    tree = cKDTree(cb_coords_rad)
    distances_rad, indices = tree.query(eu_coords_rad, distance_upper_bound=50 / 6371.0)

    matched_count = 0
    matched_ids = []
    used_at_indices = set()
    used_eu_ids = set()

    for i, (dist_rad, idx) in enumerate(zip(distances_rad, indices)):
        if not np.isfinite(dist_rad):
            continue

        if idx in used_at_indices:
            continue  # Already used this AT bus

        bus1_id = buses1_gdf.index[i]
        if bus1_id in used_eu_ids:
            continue  # Already matched this EU bus

        at_bus = cb_buses_AT.iloc[idx]

        # Assign Austrian coordinates to the matched EU bus
        network1_merged.buses.loc[bus1_id, ["x", "y"]] = at_bus.geometry.x, at_bus.geometry.y

        matched_ids.append(cb_buses_AT.index[idx])
        used_at_indices.add(idx)
        used_eu_ids.add(bus1_id)
        matched_count += 1

    print(f"✅ Matched and corrected coordinates for {matched_count} cross-border buses.")
else:
    matched_ids = []
    print("⚠️ No cross-border buses found in the Austrian network to match.")


```

```python
# # %% [markdown]
# # ### CORRECTION: Insert Unmatched Austrian Cross-Border Buses

# # %%
# # This is the corrected code block.
# # It ensures connections are only made to foreign buses within a reasonable radius.
# unmatched = cb_buses_AT[~cb_buses_AT.index.isin(matched_ids)]
# print(f"Inserting {len(unmatched)} unmatched cross-border buses from Austria.")

# # Define a reasonable search radius (e.g., in degrees, ~1.5 degrees is approx. 167 km)
# search_radius = 7 #3 #1.5 

# for idx, row in unmatched.iterrows():
#     new_bus_id = f"AT_cb_{idx}"
#     new_bus_geom = row.geometry
#     x, y = new_bus_geom.x, new_bus_geom.y
#     v_nom = row.v_nom

#     # Add the new bus to the merged network
#     network1_merged.add("Bus", new_bus_id, x=x, y=y, v_nom=v_nom, carrier="AC", country="AT")
#     print(f"  → Added bus {new_bus_id}")

#     # --- Connect to the nearest Austrian bus ---
#     austria_buses = buses1_gdf[buses1_gdf.country == "AT"].copy()
#     if not austria_buses.empty:
#         austria_buses["dist"] = austria_buses.geometry.distance(new_bus_geom)
#         near_at = austria_buses.sort_values("dist").iloc[0]
        
#         ### Fix x and r
#         length_km = near_at.dist * 111
#         r, x = estimate_line_impedance(length_km, v_nom)

#         network1_merged.add(
#             "Line", 
#             f"{new_bus_id}-{near_at.name}", 
#             bus0=new_bus_id, 
#             bus1=near_at.name, 
#             length=near_at.dist * 111, # Approx conversion from degrees to km
#             s_nom=1000, # Placeholder capacity
#             x=x, r=r
#         )

#     # # --- Connect to the nearest FOREIGN bus WITHIN the search radius ---
#     # foreign_buses = buses1_gdf[buses1_gdf.country != "AT"].copy()
    
#     # # Create a bounding box for a faster initial filter
#     # nearby_foreign_buses = foreign_buses.cx[
#     #     new_bus_geom.x - search_radius:new_bus_geom.x + search_radius,
#     #     new_bus_geom.y - search_radius:new_bus_geom.y + search_radius,
#     # ].copy()

#     # if not nearby_foreign_buses.empty:
#     #     # Calculate precise distance for the filtered subset
#     #     nearby_foreign_buses["dist"] = nearby_foreign_buses.geometry.distance(new_bus_geom)
        
#     #     # Filter again to create a circular radius
#     #     nearby_foreign_buses = nearby_foreign_buses[nearby_foreign_buses.dist < search_radius]

#     #     if not nearby_foreign_buses.empty:
#     #         near_foreign = nearby_foreign_buses.sort_values("dist").iloc[0]
#     #         network1_merged.add(
#     #             "Line",
#     #             f"{new_bus_id}-{near_foreign.name}",
#     #             bus0=new_bus_id,
#     #             bus1=near_foreign.name,
#     #             length=near_foreign.dist * 111, # Approx conversion from degrees to km
#     #             s_nom=1000 # Placeholder capacity
#     #         )
#     #         print(f"  → Connected {new_bus_id} to foreign bus {near_foreign.name}")
#     #     else:
#     #         print(f"  → WARNING: No foreign bus found within {search_radius}° for {new_bus_id}")
#     # else:
#     #     print(f"  → WARNING: No foreign bus found within the bounding box for {new_bus_id}")

#     # --- Connect to the nearest matching FOREIGN bus ---
#     # Translate foreign_country code to actual country code in the data, if needed
#     # Assuming foreign_country already matches values in `buses1_gdf["country"]`
#     foreign_code = row["foreign_country"]
#     foreign_buses = buses1_gdf[buses1_gdf.country == foreign_code].copy()

#     # Apply bounding box for spatial pre-filter
#     nearby_foreign_buses = foreign_buses.cx[
#         x - search_radius : x + search_radius,
#         y - search_radius : y + search_radius
#     ].copy()

#     if not nearby_foreign_buses.empty:
#         # Compute precise distance
#         nearby_foreign_buses["dist"] = nearby_foreign_buses.geometry.distance(new_bus_geom)

#         # Restrict to true radius
#         nearby_foreign_buses = nearby_foreign_buses[nearby_foreign_buses.dist < search_radius]

#         if not nearby_foreign_buses.empty:
#             near_foreign = nearby_foreign_buses.sort_values("dist").iloc[0]

#             ### Fix x and r
#             length_km = near_foreign.dist * 111
#             r, x = estimate_line_impedance(length_km, v_nom)

#             network1_merged.add(
#                 "Line",
#                 f"{new_bus_id}-{near_foreign.name}",
#                 bus0=new_bus_id,
#                 bus1=near_foreign.name,
#                 length=near_foreign.dist * 111,
#                 s_nom=1000,
#                 x=x, r=r
#             )
#             print(f"  → Connected {new_bus_id} to {foreign_code} bus {near_foreign.name}")
#         else:
#             print(f"  → WARNING: No {foreign_code} bus found within {search_radius}° for {new_bus_id}")
#     else:
#         print(f"  → WARNING: No {foreign_code} bus found in bounding box for {new_bus_id}")
    
#         # --- Fallback if no matching foreign bus found ---
#         if foreign_buses.empty or nearby_foreign_buses.empty:
#             print(f"    → WARNING: No {foreign_code} bus found for {new_bus_id}, trying fallback match...")

#             # Look for any foreign bus (not AT)
#             all_foreign = buses1_gdf[buses1_gdf.country != "AT"].copy()
#             all_foreign["dist"] = all_foreign.geometry.distance(geom)

#             fallback = all_foreign.sort_values("dist").iloc[0]
#             fallback_dist_km = fallback["dist"] * 111

#             r, x = estimate_line_impedance(fallback_dist_km, v_nom)
#             network1_merged.add(
#                 "Line",
#                 f"{new_bus_id}-{fallback.name}",
#                 bus0=new_bus_id, bus1=fallback.name,
#                 length=fallback_dist_km,
#                 s_nom=1000,
#                 r=r, x=x
#             )
#             print(f"    → Fallback: Connected {new_bus_id} to {fallback.country} bus {fallback.name} ({fallback_dist_km:.1f} km)")


unmatched = cb_buses_AT[~cb_buses_AT.index.isin(matched_ids)]
print(f"Inserting {len(unmatched)} unmatched cross-border buses from Austria.")

search_radius = 7  # in degrees ~780 km, adjust as needed

for idx, row in unmatched.iterrows():
    new_bus_id = f"AT_cb_{idx}"
    new_bus_geom = row.geometry
    x, y = new_bus_geom.x, new_bus_geom.y
    v_nom = row.v_nom

    network1_merged.add("Bus", new_bus_id, x=x, y=y, v_nom=v_nom, carrier="AC", country="AT")
    print(f"  → Added bus {new_bus_id}")

    # --- Connect to nearest Austrian bus ---
    austria_buses = buses1_gdf[buses1_gdf.country == "AT"].copy()
    if not austria_buses.empty:
        austria_buses["dist"] = austria_buses.geometry.distance(new_bus_geom)
        near_at = austria_buses.sort_values("dist").iloc[0]
        dist_km = near_at["dist"] * 111
        r, x = estimate_line_impedance(dist_km, v_nom)

        network1_merged.add(
            "Line",
            f"{new_bus_id}-{near_at.name}",
            bus0=new_bus_id, bus1=near_at.name,
            length=dist_km, s_nom=1000, r=r, x=x
        )

    # --- Try connecting to foreign bus by country code ---
    foreign_code = row["foreign_country"]
    foreign_buses = buses1_gdf[buses1_gdf.country == foreign_code].copy()
    nearby_foreign_buses = foreign_buses.cx[
        x - search_radius : x + search_radius,
        y - search_radius : y + search_radius
    ].copy()

    connected_foreign = False

    if not nearby_foreign_buses.empty:
        nearby_foreign_buses["dist"] = nearby_foreign_buses.geometry.distance(new_bus_geom)
        nearby_foreign_buses = nearby_foreign_buses[nearby_foreign_buses.dist < search_radius]

        if not nearby_foreign_buses.empty:
            near_foreign = nearby_foreign_buses.sort_values("dist").iloc[0]
            dist_km = near_foreign["dist"] * 111
            r, x = estimate_line_impedance(dist_km, v_nom)

            network1_merged.add(
                "Line",
                f"{new_bus_id}-{near_foreign.name}",
                bus0=new_bus_id, bus1=near_foreign.name,
                length=dist_km, s_nom=1000, r=r, x=x
            )
            print(f"  → Connected {new_bus_id} to {foreign_code} bus {near_foreign.name}")
            connected_foreign = True

    # --- Fallback: connect to any nearest foreign bus ---
    if not connected_foreign:
        print(f"  → Fallback: No matching {foreign_code} bus found for {new_bus_id}, using nearest foreign bus...")
        all_foreign = buses1_gdf[buses1_gdf.country != "AT"].copy()
        all_foreign["dist"] = all_foreign.geometry.distance(new_bus_geom)
        fallback = all_foreign.sort_values("dist").iloc[0]
        dist_km = fallback["dist"] * 111
        r, x = estimate_line_impedance(dist_km, v_nom)

        network1_merged.add(
            "Line",
            f"{new_bus_id}-{fallback.name}",
            bus0=new_bus_id, bus1=fallback.name,
            length=dist_km, s_nom=1000, r=r, x=x
        )
        print(f"    → Fallback: Connected {new_bus_id} to {fallback.country} bus {fallback.name} ({dist_km:.1f} km)")

```

```python
# %% [markdown]
# ### Split Existing Lines if Necessary

# %%
# Split lines between near_at and near_foreign buses if already connected
print("\nChecking for existing parallel lines to split...")

# Update the buses GeoDataFrame with the newly added buses
all_buses_merged = network1_merged.buses.copy()
all_buses_merged["geometry"] = gpd.points_from_xy(all_buses_merged.x, all_buses_merged.y)
all_buses_merged_gdf = gpd.GeoDataFrame(all_buses_merged, geometry="geometry", crs="EPSG:4326")

# New variables for tracking
lines_found = 0
lines_split = 0
lines_removed = 0

for idx, row in unmatched.iterrows():
    new_bus_id = f"AT_cb_{idx}"
    
    # Find the line connecting the new bus to a foreign bus
    new_line_to_foreign = network1_merged.lines[
        (network1_merged.lines.bus0 == new_bus_id) & (network1_merged.lines.bus1.str.contains('AT_cb_') == False)
    ]
    if new_line_to_foreign.empty:
        continue # Skip if no foreign connection was made

    # Get the Austrian and foreign buses this new bus is connected to
    connected_at_bus_name = network1_merged.lines.bus1[network1_merged.lines.bus0 == new_bus_id].iloc[0]
    connected_foreign_bus_name = new_line_to_foreign.bus1.iloc[0]

    # Check if a direct line already exists between the Austrian and foreign buses
    lines_df = network1.lines # Check in the *original* network
    mask = ((lines_df.bus0 == connected_at_bus_name) & (lines_df.bus1 == connected_foreign_bus_name)) | \
           ((lines_df.bus0 == connected_foreign_bus_name) & (lines_df.bus1 == connected_at_bus_name))
    
    # Iterate over any original lines that directly connect the two
    for line_id, old_line in lines_df[mask].iterrows():
        print(f"  → Found pre-existing line {line_id} between {connected_at_bus_name} and {connected_foreign_bus_name}.")
        print(f"  → It will be split by inserting {new_bus_id}.")

        # Increment the lines_found counter
        lines_found += 1
        
        # In the merged network, remove the original line if it exists
        if line_id in network1_merged.lines.index:
            network1_merged.remove("Line", line_id)
            lines_removed += 1
        
        # Also remove the two lines that were just added speculatively
        network1_merged.remove("Line", f"{new_bus_id}-{connected_at_bus_name}")
        network1_merged.remove("Line", f"{new_bus_id}-{connected_foreign_bus_name}")
        
        # Add two new lines to split the original one with the new bus
        network1_merged.add("Line", f"{connected_at_bus_name}-{new_bus_id}", bus0=connected_at_bus_name, bus1=new_bus_id, length=old_line.length / 2, s_nom=old_line.s_nom)
        network1_merged.add("Line", f"{new_bus_id}-{connected_foreign_bus_name}", bus0=new_bus_id, bus1=connected_foreign_bus_name, length=old_line.length / 2, s_nom=old_line.s_nom)
        print(f"  → Line {line_id} successfully split.")
```

```python
print("\nChecking for existing direct lines to split via new cross-border buses...")

# Prepare merged GeoDataFrame of buses (to ensure locations are available)
all_buses_merged = network1_merged.buses.copy()
all_buses_merged["geometry"] = gpd.points_from_xy(all_buses_merged.x, all_buses_merged.y)
all_buses_merged_gdf = gpd.GeoDataFrame(all_buses_merged, geometry="geometry", crs="EPSG:4326")

# Tracking counters
lines_found = 0
lines_removed = 0
lines_split = 0

unmatched = cb_buses_AT[~cb_buses_AT.index.isin(matched_ids)]


for idx, row in unmatched.iterrows():
    new_bus_id = f"AT_cb_{idx}"
    
    # Foreign connection line
    foreign_line_mask = (
        (network1_merged.lines.bus0 == new_bus_id) & 
        (~network1_merged.lines.bus1.str.startswith("AT_cb_"))
    )
    if not foreign_line_mask.any():
        continue  # Skip if no foreign connection made

    # Extract connected buses
    foreign_line = network1_merged.lines[foreign_line_mask].iloc[0]
    connected_foreign_bus = foreign_line.bus1

    # Austrian connection
    try:
        at_line = network1_merged.lines[(network1_merged.lines.bus0 == new_bus_id) & 
                                        (network1_merged.lines.bus1 != connected_foreign_bus)].iloc[0]
        connected_at_bus = at_line.bus1
    except IndexError:
        print(f"  → Skipping {new_bus_id}: Could not find connected Austrian bus.")
        continue

    # Check original network for redundant direct line
    lines_df = network1.lines
    direct_line_mask = (
        ((lines_df.bus0 == connected_at_bus) & (lines_df.bus1 == connected_foreign_bus)) |
        ((lines_df.bus0 == connected_foreign_bus) & (lines_df.bus1 == connected_at_bus))
    )

    existing_lines = lines_df[direct_line_mask]
    if existing_lines.empty:
        continue  # No direct line to split

    for line_id, old_line in existing_lines.iterrows():
        print(f"  → Found redundant line {line_id} between {connected_at_bus} and {connected_foreign_bus}. Will split via {new_bus_id}.")
        lines_found += 1

        # Remove line from merged network
        if line_id in network1_merged.lines.index:
            network1_merged.remove("Line", line_id)
            lines_removed += 1

        # Remove speculative lines (if already inserted)
        for link_id in [
            f"{new_bus_id}-{connected_at_bus}",
            f"{connected_at_bus}-{new_bus_id}",
            f"{new_bus_id}-{connected_foreign_bus}",
            f"{connected_foreign_bus}-{new_bus_id}"
        ]:
            if link_id in network1_merged.lines.index:
                network1_merged.remove("Line", link_id)

        # Insert two new lines
        half_length = old_line.length / 2 if pd.notna(old_line.length) else 50
        r, x = estimate_line_impedance(half_length, old_line.v_nom)

        network1_merged.add(
            "Line", f"{connected_at_bus}-{new_bus_id}",
            bus0=connected_at_bus, bus1=new_bus_id,
            length=half_length,
            s_nom=old_line.s_nom,
            r=r,
            x=x
        )
        network1_merged.add(
            "Line", f"{new_bus_id}-{connected_foreign_bus}", 
            bus0=new_bus_id, bus1=connected_foreign_bus, 
            length=half_length, s_nom=old_line.s_nom,
            r=r,
            x=x
        )
        print(f"  → Line {line_id} successfully split with {new_bus_id}.")
        lines_split += 1

print(f"\n✔ {lines_found} redundant lines found.")
print(f"✔ {lines_removed} lines removed.")
print(f"✔ {lines_split} lines split via cross-border buses.")
```

```python
print("\n🔍 Listing all remaining direct AT↔foreign lines:")

cross_border_direct_lines = []

for line_id, line in network1_merged.lines.iterrows():
    bus0, bus1 = line.bus0, line.bus1
    if "AT_cb_" in bus0 or "AT_cb_" in bus1:
        continue  # skip CB inserted links

    try:
        country0 = network1_merged.buses.loc[bus0, "country"]
        country1 = network1_merged.buses.loc[bus1, "country"]
    except KeyError:
        print(f"⚠️ Line {line_id} has invalid bus reference: {bus0}, {bus1}")
        continue

    if country0 == "AT" and country1 != "AT":
        cross_border_direct_lines.append((line_id, bus0, country0, bus1, country1))
    elif country1 == "AT" and country0 != "AT":
        cross_border_direct_lines.append((line_id, bus1, country1, bus0, country0))

print(f"\n✳️ Found {len(cross_border_direct_lines)} direct AT↔foreign lines (not via AT_cb_):")

for line in cross_border_direct_lines:
    print(f"  - Line {line[0]}: {line[1]} ({line[2]}) ↔ {line[3]} ({line[4]})")
```

```python
print("\n🪛 Splitting remaining direct AT↔foreign lines using fallback midpoint logic...")

for i, (line_id, at_bus, at_country, foreign_bus, foreign_country) in enumerate(cross_border_direct_lines):
    line = network1_merged.lines.loc[line_id]

    # Remove the direct line
    network1_merged.remove("Line", line_id)

    # Get coordinates of both ends
    x0, y0 = network1_merged.buses.loc[at_bus, ["x", "y"]]
    x1, y1 = network1_merged.buses.loc[foreign_bus, ["x", "y"]]

    mid_x, mid_y = (x0 + x1) / 2, (y0 + y1) / 2
    length_total = line.length or 100
    half_length = length_total / 2
    v_nom = line.v_nom or 220
    s_nom = line.s_nom or 1000

    # Estimate impedance
    r, x = estimate_line_impedance(half_length, v_nom)

    cb_bus_id = f"AT_cb_fallback_{i}"
    network1_merged.add("Bus", cb_bus_id, x=mid_x, y=mid_y, v_nom=v_nom, country="AT", carrier="AC")

    # Add two split lines
    network1_merged.add("Line", f"{at_bus}-{cb_bus_id}", bus0=at_bus, bus1=cb_bus_id,
                        length=half_length, s_nom=s_nom, r=r, x=x)
    network1_merged.add("Line", f"{cb_bus_id}-{foreign_bus}", bus0=cb_bus_id, bus1=foreign_bus,
                        length=half_length, s_nom=s_nom, r=r, x=x)

    print(f"  🔧 Replaced {line_id}: {at_bus} ↔ {foreign_bus} with {cb_bus_id}")

print("\n✅ All remaining direct lines now split via fallback CB buses.")
```
