## Summing up residual loads
Residual loads are present in the CSV files as normalized units. Therefore, the time series are summed by node and multiplied by the total consumption of Austria to obtain the unit MW. Afterwards, the summed time series is distributed back to the Austrian network nodes.

```python
import pandas as pd
from pathlib import Path

# Folder with time series
folder = Path(f"{snakemake.input.loads_dir}/NIP2030")  # or NIP2040

# Only CSV files (e.g. 'Nachfrage.csv', 'PV.csv', 'Wind.csv', 'Export.csv', etc.)
data_files = [f for f in folder.glob("*.csv") if f.name != "Metadata.csv"]

# Initialize empty DataFrame for total sum
total_relative_profile = None

for f in data_files:
    df = pd.read_csv(f, index_col=0, delimiter=";", parse_dates=True)
    # For Export.csv, the time profile starts from row 1, row 0 is mapping
    if f.name.lower().startswith("export"):
        df = df.drop(index=df.index[0])  # Remove mapping row
    df = df.astype(float)
    summed = df.sum(axis=1)  # Time series: sum of all columns for each hour
    # Reset index (so each summed time series is unique)
    summed.index = pd.RangeIndex(start=0, stop=len(summed))
    if total_relative_profile is None:
        total_relative_profile = summed
    else:
        total_relative_profile += summed
```

```python
# Total consumption as time series (e.g. from PyPSA or Statistik Austria)
real_total_load_mw = 5500  # Example: average consumption in MW
residual_load_mw = total_relative_profile / total_relative_profile.mean() * real_total_load_mw
```
