```python

# --- Section 2: Helper Functions ---

def estimate_line_impedance(length_km, v_nom_kv):
    if v_nom_kv >= 380:
        r_per_km, x_per_km = 0.03, 0.3
    elif v_nom_kv >= 220:
        r_per_km, x_per_km = 0.05, 0.4
    elif v_nom_kv >= 110:
        r_per_km, x_per_km = 0.1, 0.6
    else:
        r_per_km, x_per_km = 0.2, 0.8
    return r_per_km * length_km, x_per_km * length_km


def haversine_km(latlon1, latlon2):
    lat1, lon1 = np.radians(latlon1)
    lat2, lon2 = np.radians(latlon2)
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    c = 2 * np.arcsin(np.sqrt(a))
    return 6371.0 * c


def plot_cb_matches(cb_gdf, matched_gdf, unmatched_gdf=None, borders=None):
    fig, ax = plt.subplots(figsize=(10, 10))
    if borders is not None:
        borders.plot(ax=ax, facecolor='none', edgecolor='black')
    cb_gdf.plot(ax=ax, color='red', markersize=20, label='CB Buses (AT)')
    matched_gdf.plot(ax=ax, color='green', markersize=15, label='Matched Buses (PyPSA-Eur)')
    if unmatched_gdf is not None:
        unmatched_gdf.plot(ax=ax, color='orange', markersize=25, label='Unmatched CB Buses')
    plt.legend()
    plt.title("Cross-Border Bus Matching Overview")
    plt.grid(True)
    plt.show()

```

```python
# Section 6: Hybrid CB Matching
""" Match verified CB buses from networkAT to candidate buses in network1 using:

KDTree for speed

attribute filtering (voltage, country mismatch)

fallback handling if no match found """

from scipy.spatial import cKDTree
import numpy as np

def match_cb_buses_to_pypsa(cb_buses, buses1_gdf, max_km=50, voltage_tol=0.1, verbose=True):
    """
    Matches verified cross-border buses to PyPSA-Eur buses based on proximity and voltage compatibility.
    
    Parameters
    ----------
    cb_buses : GeoDataFrame
        Verified CB buses (from networkAT)
    buses1_gdf : GeoDataFrame
        PyPSA-Eur buses
    max_km : float
        Maximum allowed haversine distance (in km)
    voltage_tol : float
        Relative voltage tolerance (e.g. 0.1 = ±10%)
    verbose : bool
        Print debug info

    Returns
    -------
    matched : GeoDataFrame with matched_bus_id, distance_km, v_nom_diff
    unmatched : GeoDataFrame
    """

    # Convert to radians for haversine KDTree
    def to_radians(df): return np.radians(df[["y", "x"]].values)

    # Prepare KDTree
    candidates = buses1_gdf[buses1_gdf.country != "AT"].copy()
    tree = cKDTree(to_radians(candidates))
    cb_coords = to_radians(cb_buses)

    # Results
    matched_rows = []
    unmatched = []

    for i, cb_row in cb_buses.iterrows():
        cb_point = cb_row.geometry
        cb_v = cb_row.v_nom
        cb_coord_rad = cb_coords[list(cb_buses.index).index(i)]
        foreign_code = cb_row["foreign_country"]

        # Query nearest N neighbors (max 10)
        dists_rad, idxs = tree.query(cb_coord_rad, k=10, distance_upper_bound=max_km / 6371.0)

        found_match = False
        for dist_rad, idx in zip(dists_rad, idxs):
            if not np.isfinite(dist_rad): continue
            candidate = candidates.iloc[idx]
            cand_v = candidate.v_nom
            if cand_v <= 0: continue

            # Check voltage similarity
            rel_diff = abs(cand_v - cb_v) / cb_v
            if rel_diff > voltage_tol: continue

            # Save match
            match = cb_row.copy()
            match["matched_bus_id"] = candidate.name
            match["distance_km"] = dist_rad * 6371
            match["v_nom_diff"] = cand_v - cb_v
            matched_rows.append(match)
            found_match = True

            if verbose:
                print(f"✅ Match: {i} → {candidate.name} ({match['distance_km']:.2f} km, Δv_nom={match['v_nom_diff']:.1f})")
            break

        if not found_match:
            unmatched.append(cb_row)
            if verbose:
                print(f"❌ No match found for {i}")

    matched_gdf = gpd.GeoDataFrame(matched_rows, geometry="geometry", crs=cb_buses.crs)
    unmatched_gdf = gpd.GeoDataFrame(unmatched, geometry="geometry", crs=cb_buses.crs)

    return matched_gdf, unmatched_gdf
```

```python
def insert_cb_bus_with_fallback(cb_row, network, buses1_gdf, verbose=True):
    """
    Inserts a new cross-border bus into the PyPSA network and connects it to
    the nearest Austrian and foreign buses with impedance-aware lines.

    Parameters
    ----------
    cb_row : pd.Series (one row from verified_cb_buses)
    network : pypsa.Network (modifies in-place)
    buses1_gdf : GeoDataFrame of buses from network1
    verbose : bool
    """

    new_bus_id = f"AT_cb_{cb_row.name}"
    x, y = cb_row.geometry.x, cb_row.geometry.y
    v_nom = cb_row.v_nom
    foreign_code = cb_row.foreign_country

    # --- Add the CB bus ---
    network.add("Bus", new_bus_id, x=x, y=y, v_nom=v_nom, carrier="AC", country="AT")
    if verbose:
        print(f"➕ Inserted fallback CB bus: {new_bus_id}")

    # --- Connect to nearest Austrian bus ---
    austrian_buses = buses1_gdf[buses1_gdf.country == "AT"].copy()
    austrian_buses["dist"] = austrian_buses.geometry.distance(cb_row.geometry)
    if not austrian_buses.empty:
        near_at = austrian_buses.sort_values("dist").iloc[0]
        dist_km = near_at["dist"] * 111
        r, x = estimate_line_impedance(dist_km, v_nom)

        network.add("Line", f"{new_bus_id}-{near_at.name}", bus0=new_bus_id, bus1=near_at.name,
                    length=dist_km, s_nom=1000, r=r, x=x)
        if verbose:
            print(f"   ↔ Connected to AT bus {near_at.name} ({dist_km:.1f} km)")

    # --- Connect to nearest FOREIGN bus (prefer matching country) ---
    foreign_buses = buses1_gdf[buses1_gdf.country == foreign_code].copy()
    if foreign_buses.empty:
        foreign_buses = buses1_gdf[buses1_gdf.country != "AT"].copy()  # fallback to any foreign
    foreign_buses["dist"] = foreign_buses.geometry.distance(cb_row.geometry)
    if not foreign_buses.empty:
        near_foreign = foreign_buses.sort_values("dist").iloc[0]
        dist_km = near_foreign["dist"] * 111
        r, x = estimate_line_impedance(dist_km, v_nom)

        network.add("Line", f"{new_bus_id}-{near_foreign.name}", bus0=new_bus_id, bus1=near_foreign.name,
                    length=dist_km, s_nom=1000, r=r, x=x)
        if verbose:
            print(f"   ↔ Connected to {near_foreign.country} bus {near_foreign.name} ({dist_km:.1f} km)")
```

```python
def remove_redundant_direct_lines(network, cb_prefix="AT_cb_", verbose=True):
    """
    Removes direct AT↔foreign lines that are now bridged via inserted CB buses.

    Parameters
    ----------
    network : pypsa.Network
        The modified PyPSA network (in-place)
    cb_prefix : str
        Prefix for identifying inserted CB buses
    verbose : bool
        Print match and removal info
    """
    cb_buses = [bus for bus in network.buses.index if bus.startswith(cb_prefix)]
    cb_connected_pairs = set()

    # Build set of pairs connected via CB buses
    for cb in cb_buses:
        neighbors = set()
        for _, line in network.lines.iterrows():
            if line.bus0 == cb:
                neighbors.add(line.bus1)
            elif line.bus1 == cb:
                neighbors.add(line.bus0)
        if len(neighbors) == 2:
            pair = tuple(sorted(neighbors))
            cb_connected_pairs.add(pair)

    # Now scan all lines for direct AT↔foreign connections
    lines_to_remove = []
    for line_id, line in network.lines.iterrows():
        if line.bus0.startswith(cb_prefix) or line.bus1.startswith(cb_prefix):
            continue  # skip CB-based lines

        b0_country = network.buses.loc[line.bus0, "country"]
        b1_country = network.buses.loc[line.bus1, "country"]

        if b0_country == "AT" and b1_country != "AT":
            pair = tuple(sorted([line.bus0, line.bus1]))
        elif b1_country == "AT" and b0_country != "AT":
            pair = tuple(sorted([line.bus0, line.bus1]))
        else:
            continue

        if pair in cb_connected_pairs:
            lines_to_remove.append(line_id)
            if verbose:
                print(f"❌ Removed direct line {line_id}: {pair[0]} ↔ {pair[1]}")

    # Remove lines
    for lid in lines_to_remove:
        network.remove("Line", lid)

    print(f"✅ Removed {len(lines_to_remove)} redundant direct AT↔foreign lines.")
```

```python
def reroute_lines_from_non_cb_at_buses(network, verified_cb_buses, cb_prefix="AT_cb_", max_cb_distance_km=30, reroute=True, verbose=True):
    """
    Removes remaining AT↔foreign lines where the AT bus is not a known CB and is far from any verified CB bus.

    Parameters
    ----------
    network : pypsa.Network (in-place)
    verified_cb_buses : GeoDataFrame (from AT network)
    cb_prefix : str
        Prefix of inserted CB buses
    max_cb_distance_km : float
        Distance threshold to classify "far from CB"
    reroute : bool
        Whether to reconnect to the nearest CB
    verbose : bool
        Print summary info
    """
    removed, rerouted = 0, 0

    buses = network.buses.copy()
    buses_gdf = gpd.GeoDataFrame(buses, geometry=gpd.points_from_xy(buses.x, buses.y), crs="EPSG:4326")

    # Build CB bus sets
    known_cb_ids = set(verified_cb_buses.index)
    inserted_cb_buses = buses_gdf[buses_gdf.index.str.startswith(cb_prefix)].copy()

    # Precompute verified CB geometries
    cb_tree = cKDTree(np.radians(verified_cb_buses[["y", "x"]].values))

    for line_id, line in network.lines.iterrows():
        b0, b1 = line.bus0, line.bus1
        try:
            c0, c1 = buses.loc[b0, "country"], buses.loc[b1, "country"]
        except KeyError:
            continue

        if c0 == "AT" and c1 != "AT":
            at_bus, foreign_bus = b0, b1
        elif c1 == "AT" and c0 != "AT":
            at_bus, foreign_bus = b1, b0
        else:
            continue  # not a CB line

        # Skip if AT bus is a known CB or startswith AT_cb_
        if at_bus in known_cb_ids or at_bus.startswith(cb_prefix):
            continue

        # Check distance to nearest verified CB
        at_geom = buses_gdf.loc[at_bus].geometry
        at_rad = np.radians([[at_geom.y, at_geom.x]])
        dist_rad, _ = cb_tree.query(at_rad)
        dist_km = dist_rad[0] * 6371

        if dist_km > max_cb_distance_km:
            network.remove("Line", line_id)
            removed += 1
            if verbose:
                print(f"❌ Removed: {line_id} (AT bus {at_bus} is {dist_km:.1f} km from any CB)")

            if reroute and not inserted_cb_buses.empty:
                inserted_cb_buses["dist"] = inserted_cb_buses.geometry.distance(at_geom)
                nearest_cb = inserted_cb_buses.sort_values("dist").iloc[0]
                d_km = nearest_cb["dist"] * 111
                r, x = estimate_line_impedance(d_km, line.v_nom or 220)

                network.add("Line", f"{nearest_cb.name}-{foreign_bus}",
                            bus0=nearest_cb.name,
                            bus1=foreign_bus,
                            length=d_km, s_nom=line.s_nom or 1000,
                            r=r, x=x)
                rerouted += 1
                if verbose:
                    print(f"   ↪️ Rerouted to {nearest_cb.name} ({d_km:.1f} km)")

    print(f"\n✅ Removed {removed} improperly connected AT↔foreign lines.")
    if reroute:
        print(f"✅ Inserted {rerouted} rerouted connections.")
```

```python
from shapely.geometry import Point
from shapely.strtree import STRtree
from collections import Counter

from scipy.spatial import cKDTree
import numpy as np

def find_spatial_duplicates(buses_gdf, radius_km=1):
    """
    Finds sets of buses within a certain radius of each other.

    Parameters
    ----------
    buses_gdf : GeoDataFrame with EPSG:4326
    radius_km : float
        Search radius in kilometers

    Returns
    -------
    List of lists, each a group of duplicate bus indices
    """
    coords_deg = np.radians(buses_gdf[["y", "x"]].values)  # lat/lon → radians
    tree = cKDTree(coords_deg)
    radius_rad = radius_km / 6371.0  # convert km to radians

    groups = []
    seen = set()

    for i, coord in enumerate(coords_deg):
        if i in seen:
            continue
        idxs = tree.query_ball_point(coord, r=radius_rad)
        idxs = [j for j in idxs if j != i]
        if idxs:
            group = [buses_gdf.index[i]] + [buses_gdf.index[j] for j in idxs]
            groups.append(group)
            seen.update(idxs + [i])

    return groups


def report_cb_stats(network, cb_prefix="AT_cb_", duplicate_radius_km=1):
    """
    Prints cross-border bus and line statistics and detects spatially close duplicate buses.

    Parameters
    ----------
    network : pypsa.Network
        The network to analyze
    cb_prefix : str
        Prefix used for inserted cross-border buses
    duplicate_radius_km : float
        Radius in km to search for bus duplicates (using geometry proximity)
    """
    # Count CB buses
    cb_buses = [b for b in network.buses.index if b.startswith(cb_prefix)]
    print(f"🔢 Cross-border buses ({cb_prefix}): {len(cb_buses)}")

    # Count CB lines
    cb_line_counts = Counter()
    for _, line in network.lines.iterrows():
        c0 = network.buses.loc[line.bus0, "country"]
        c1 = network.buses.loc[line.bus1, "country"]
        if c0 == "AT" and c1 != "AT":
            cb_line_counts[c1] += 1
        elif c1 == "AT" and c0 != "AT":
            cb_line_counts[c0] += 1

    total_cb_lines = sum(cb_line_counts.values())
    print(f"🔢 Total cross-border lines: {total_cb_lines}")
    print("   Breakdown by foreign country:")
    for country, count in cb_line_counts.items():
        print(f"   → AT ↔ {country}: {count}")

    # Detect spatial duplicates (within radius)
    buses = network.buses.copy()
    buses_gdf = gpd.GeoDataFrame(
        buses, geometry=gpd.points_from_xy(buses.x, buses.y), crs="EPSG:4326"
    )

    geometries = buses_gdf.geometry.to_list()
    tree = STRtree(geometries)

    duplicate_sets = find_spatial_duplicates(buses_gdf, radius_km=duplicate_radius_km)

    if duplicate_sets:
        print(f"\n⚠️ Found {len(duplicate_sets)} spatial duplicate bus groups (within {duplicate_radius_km} km):")
        for group in duplicate_sets:
            print(f"   • {group}")
    else:
        print(f"\n✅ No spatial duplicates within {duplicate_radius_km} km found.")
```

```python
import matplotlib.pyplot as plt

def plot_cb_voltage_clusters(cb_buses_gdf, borders=None, cluster_radius_km=5, title="Cross-Border Voltage Clusters"):
    """
    Plots CB substations with multiple voltage levels as spatial clusters.

    Parameters
    ----------
    cb_buses_gdf : GeoDataFrame
        GeoDataFrame of verified_cb_buses
    borders : GeoDataFrame
        Optional: country borders
    cluster_radius_km : float
        Radius (km) within which buses are considered one cluster
    title : str
        Plot title
    """
    # Use cKDTree to group by proximity
    coords_rad = np.radians(cb_buses_gdf[["y", "x"]].values)
    tree = cKDTree(coords_rad)
    radius_rad = cluster_radius_km / 6371

    groups = []
    seen = set()
    for i, coord in enumerate(coords_rad):
        if i in seen:
            continue
        idxs = tree.query_ball_point(coord, radius_rad)
        idxs = [j for j in idxs if j != i]
        if idxs:
            group = [cb_buses_gdf.index[i]] + [cb_buses_gdf.index[j] for j in idxs]
            groups.append(group)
            seen.update(idxs + [i])

    # Plot
    fig, ax = plt.subplots(figsize=(10, 10))
    if borders is not None:
        borders.plot(ax=ax, facecolor="none", edgecolor="gray", linewidth=0.5)

    # Color by voltage
    voltages = sorted(cb_buses_gdf.v_nom.unique())
    cmap = plt.get_cmap("tab10", len(voltages))
    color_map = {v: cmap(i) for i, v in enumerate(voltages)}

    # Plot clusters
    for group in groups:
        cluster = cb_buses_gdf.loc[group]
        for v_nom, gdf in cluster.groupby("v_nom"):
            gdf.plot(ax=ax, markersize=60, marker="o", color=color_map[v_nom], label=f"{int(v_nom)} kV")

    # Plot remaining ungrouped (if any)
    ungrouped = cb_buses_gdf[~cb_buses_gdf.index.isin([i for group in groups for i in group])]
    if not ungrouped.empty:
        for v_nom, gdf in ungrouped.groupby("v_nom"):
            gdf.plot(ax=ax, markersize=30, marker="x", color=color_map[v_nom], label=f"{int(v_nom)} kV (ungrouped)")

    # Finish plot
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(dict(zip(labels, handles)).values(), dict(zip(labels, handles)).keys())
    ax.set_title(title)
    ax.set_xlim(7, 18)
    ax.set_ylim(45.5, 49.5)
    ax.grid(True)
    plt.tight_layout()
    plt.show()
```
