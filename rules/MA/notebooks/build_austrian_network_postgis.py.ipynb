````python
import pypsa
import geopandas as gpd
from shapely.geometry import LineString, MultiLineString, Point
import pandas as pd

# Initialize PyPSA network
network = pypsa.Network()

# ---- Add Buses ----
for _, row in buses_gdf.iterrows():
    bus_id = str(row["id"])
    geom = row["geom"]
    if not isinstance(geom, Point):
        continue  # skip if geometry isn't valid
    
    network.add("Bus",
                name=bus_id,
                x=geom.x,
                y=geom.y,
                v_nom=row.get("v_ref", 110),
                type=str(row.get("type", "substation")),
                substation_name=row.get("name", None),
                s_nom=row.get("s_nom", None),
                build_date=row.get("comm", None),
                lifetime=(row["decomm"].year - row["comm"].year) if pd.notna(row.get("comm")) and pd.notna(row.get("decomm")) else None
                )

# ---- Add Lines ----
for _, row in lines_gdf.iterrows():
    line_id = str(row["id"])
    from_bus = str(row["from_node"])
    to_bus = str(row["to_node"])
    
    geom = row["geom"]
    if isinstance(geom, MultiLineString):
        # Flatten MultiLineString to a single LineString for PyPSA
        coords = []
        for line in geom.geoms:
            coords.extend(line.coords)
        line_geom = LineString(coords)
    elif isinstance(geom, LineString):
        line_geom = geom
    else:
        continue  # Skip invalid geometries

    network.add("Line",
                name=f"Line_{line_id}",
                bus0=from_bus,
                bus1=to_bus,
                x=row.get("x", 0),
                r=row.get("r", 0.01),  # fallback if missing
                s_nom=row.get("s_max", 100),
                type=row.get("type", None),
                build_date=row.get("comm", None),
                capital_cost=None,  # optionally calculate based on length or s_nom
                length=row.get("length", None),
                c=row.get("c", None),
                g=0,  # conductance, set to zero unless known
                v_nom=row.get("v_ref", 110),
                i_nom=row.get("i_max", None),
                # Optional custom attributes as tags
                tags={
                    "name": row.get("name", None),
                    "decomm": row.get("decomm", None),
                    "scenario": row.get("scenario", None),
                    "references": row.get("references", None),
                })

# Optional: set name for the network
network.name = "Austrian Grid from PostGIS (MultiLineString-aware)"

# Save the network
#network.export_to_netcdf("austrian_grid.nc")
````

````python
### filter only buses with type = 3

# Step 1: Filter buses with type == 3
filtered_buses_gdf = buses_gdf[buses_gdf["type"] == 3].copy()

# Step 2: Get the IDs of the selected buses
selected_bus_ids = set(filtered_buses_gdf["id"].astype(str))

# Step 3: Filter lines where both from_node and to_node are in selected_bus_ids
filtered_lines_gdf = lines_gdf[
    lines_gdf["from_node"].astype(str).isin(selected_bus_ids) &
    lines_gdf["to_node"].astype(str).isin(selected_bus_ids)
].copy()

# Step 4: Create new PyPSA network
filtered_network = pypsa.Network()

# Add filtered buses
for _, row in filtered_buses_gdf.iterrows():
    bus_id = str(row["id"])
    geom = row["geom"]
    if not isinstance(geom, Point):
        continue

    filtered_network.add("Bus",
                         name=bus_id,
                         x=geom.x,
                         y=geom.y,
                         v_nom=row.get("v_ref", 110),
                         type=str(row.get("type", "substation")),
                         substation_name=row.get("name", None),
                         s_nom=row.get("s_nom", None),
                         build_date=row.get("comm", None),
                         lifetime=(row["decomm"].year - row["comm"].year) if pd.notna(row.get("comm")) and pd.notna(row.get("decomm")) else None
                         )

# Add matching lines
for _, row in filtered_lines_gdf.iterrows():
    line_id = str(row["id"])
    from_bus = str(row["from_node"])
    to_bus = str(row["to_node"])
    
    geom = row["geom"]
    if isinstance(geom, MultiLineString):
        coords = []
        for line in geom.geoms:
            coords.extend(line.coords)
        line_geom = LineString(coords)
    elif isinstance(geom, LineString):
        line_geom = geom
    else:
        continue

    filtered_network.add("Line",
                         name=f"Line_{line_id}",
                         bus0=from_bus,
                         bus1=to_bus,
                         x=row.get("x", 0),
                         r=row.get("r", 0.01),
                         s_nom=row.get("s_max", 100),
                         type=row.get("type", None),
                         build_date=row.get("comm", None),
                         length=row.get("length", None),
                         c=row.get("c", None),
                         g=0,
                         v_nom=row.get("v_ref", 110),
                         i_nom=row.get("i_max", None),
                         tags={
                             "name": row.get("name", None),
                             "decomm": row.get("decomm", None),
                             "scenario": row.get("scenario", None),
                             "references": row.get("references", None),
                         })
````
